#### Exercice 1 : Exploration du module math à travers le calcul d'un disque

#####  L'objectif:

L'idée ici est de se familiariser avec la bibliothèque math de Python pour automatiser des calculs géométriques simples, tout en apprenant à dompter l'affichage des chiffres après la virgule.

##### Mon analyse de l'énoncé :

En préparant l'exercice, j'ai remarqué une petite erreur dans la consigne. L'énoncé nous donne une valeur de 4.56 cm présentée comme le diamètre. Pourtant, si l'on suit les formules classiques :

Périmètre (2πr)

Aire (πr^2)

...les résultats cibles (65.33 et 28.65) ne correspondent pas au diamètre de 4.56. En réalité, on obtient exactement ces chiffres si l'on traite 4.56 comme le rayon (r).

##### Ma stratégie :

Pour que mon rapport soit complet, j'ai décidé de coder les deux versions. Cela montre que j'ai bien identifié l'écart entre le texte de la consigne et le résultat attendu.

```python
import math
#On considère r = 4.56 est le diamètre
diametre = 4.56
rayon = diametre / 2
perimetre = math.pi * diametre
aire = math.pi * (rayon ** 2)
print(f"L'aire du disque: {round(aire, 2)} \n Le perimètre du disque : {round(perimetre, 2)}")
#On considère r = 4.56 est le rayon
rayon = 4.56
perimetre = math.pi * rayon * 2
aire = math.pi * (rayon ** 2)
print(f"L'aire du disque: {round(aire, 2)} \n Le perimètre du disque : {round(perimetre, 2)}")
```

##### Observations personnelles :

* Précision :
L'utilisation de math.pi est bien plus fiable que de taper manuellement "3.14", car elle utilise la précision maximale de la machine.

* Formatage :
 La fonction round() est pratique pour obtenir un résultat propre (deux décimales).


#### Exercice 2 : Somme de fractions et PGCD (Plus Grand Commun Diviseur)

#####  L'objectif:

Réaliser une addition de trois fractions manuel (par le calcul) sans utiliser avec les modules automatiques dans un premier temps. L'idée est de retrouver la logique mathématique : mettre au même dénominateur, additionner, puis simplifier.

##### Ma stratégie de résolution :

1. Dénominateur commun : Pour simplifier, j'ai multiplié les trois dénominateurs entre eux (440×256×71).
2. Ajustement des numérateurs : Chaque numérateur est multiplié par les dénominateurs des deux autres fractions.
3. Simplification : C'est ici qu'intervient la fonction math.gcd. En divisant le numérateur et le dénominateur par leur plus grand commun diviseur, on obtient la forme irréductible.
4. Vérification : J'utilise ensuite le module fractions pour confirmer que mon calcul manuel est correct.

```python
import math
from fractions import Fraction
deno = 440 * 256 * 71
num = (217 * 256 * 71)+(101 * 440 * 71)+(440 * 256 * 86)
commun = math.gcd(num,deno)
num_irrd = num // commun
deno_irrd = deno // commun
print(f"Forme irréductible : {num_irrd}/{deno_irrd}")
f1 = Fraction(217,440)
f2 = Fraction(101,256)
f3 = Fraction(86,71)
somme_verif = f1 + f2 + f3
print (f"Résultat via le module fractions: {somme_verif}")
```

##### Points clés de l'exercice :

* Puissance du PGCD :La fonction math.gcd simplifie instantanément des nombres très élevés (ici dans les millions), ce qui rend le calcul manuel de fractions irréductibles extrêmement efficace.
* Précision absolue : En travaillant avec des entiers plutôt que des float, on élimine tout risque d'erreur d'arrondi, garantissant une exactitude mathématique parfaite.
* Vérification réussie :La correspondance exacte entre mon calcul et le module fractions confirme que la logique de mise au même dénominateur a été correctement implémentée.


#### Exercice 3 : Recherche du maximum entre trois entiers

##### L'objectif :

Concevoir un programme capable de comparer trois nombres saisis par l'utilisateur pour en extraire la valeur maximale. Cet exercice permet de mettre en pratique la récupération de données via input() et la logique des branchements conditionnels.

##### Ma logique de résolution :

* Saisie de données : Utilisation de int(input()) pour s'assurer que les entrées sont bien traitées comme des nombres entiers.
* Structure conditionnelle : J'ai utilisé une cascade de conditions if, elif et else pour tester chaque nombre par rapport aux deux autres.
* Efficacité : Cette méthode permet de trouver le résultat de manière directe, sans utiliser de fonctions prédéfinies comme max(), afin de démontrer la logique algorithmique de base.

```python
n1 = int(input("Entrer le premier entier :"))
n2 = int(input("Entrer le second entier :"))
n3 = int(input("Entrer le troisième entier :"))
if n1 > n2 and n1 > n3 :
    max_val = n1
elif n2 > n1 and n2 > n3 :
    max_val = n2
else:
    max_val = n3
print(f"{n1} {n2} {n3} ==> {max_val}")
```

##### Points clés de l'exercice :

* Validation des types : L'utilisation de int() est cruciale pour éviter que Python ne compare les entrées comme des chaînes de caractères (ce qui fausserait le résultat).
* Structure logique : Le bloc if/elif/else est efficace ici car il s'arrête dès qu'une condition est remplie, ce qui optimise légèrement l'exécution.

#### Exercice 6 : Gestion des horaires d'ouverture (Logique booléenne)

##### L'objectif :

Déterminer si un guichet est ouvert ou fermé en croisant deux variables : le jour de la semaine et l'heure. Ce problème demande une gestion précise des intervalles de temps et des exceptions du week-end.

##### Ma stratégie de résolution :

* Centralisation du résultat : J'utilise une variable booléenne ouvert initialisée à False. Le but est de ne la passer à True que si toutes les conditions d'ouverture sont réunies.
* Découpage des conditions :
   1. Semaine (Lundi à Vendredi) : Vérification des deux plages horaires (matin et après-midi).
   2. Samedi : Une condition spécifique pour n'ouvrir que le matin (8h-13h).
   3. Dimanche : Il reste fermé par défaut (grâce à l'initialisation).
* Résultat attendu : L'affichage doit être un booléen pur (Vrai ou Faux) pour correspondre aux jeux de tests.

```python
jour = int(input("Entrer le jour:"))
heure = int(input("Entrer l'heure :"))
ouvert = False
if jour in range(1,6) and ((heure >= 8 and heure <= 13) or (heure >= 14 and heure <= 17)):
    ouvert = True
if jour == 6 and heure >= 8 and heure <= 13 :
    ouvert = True
if jour == 0:
    ouvert = False
if ouvert:
    print("Vrai")
else:
    print("Faux")
```

##### Points clés de l'exercice :

* Précision des bornes : L'utilisation de comparaisons enchaînées (8 <= heure <= 13) rend le code plus lisible et plus proche du langage mathématique.
* Gestion des exceptions : Le samedi est traité à part car ses horaires diffèrent de la semaine, ce qui évite d'ouvrir le guichet par erreur le samedi après-midi.
* Robustesse : En initialisant ouvert à False, on s'assure que le programme est "sécurisé" : par défaut, tout ce qui n'est pas explicitement autorisé est considéré comme fermé.


#### Exercice 7 : Détermination des saisons (Conditions complexes)

##### L'objectif :

Identifier la saison correspondante à une date donnée (m,j). Ce problème est intéressant car il nécessite de jongler avec des intervalles de temps qui "chevauchent" deux mois différents.

##### Ma stratégie de résolution :

* Décomposition par saison : Pour chaque saison, j'ai identifié trois cas de figure :

  1. Le mois de début avec une condition sur le jour (ex: après le 22 décembre).
  2. Les mois pleins situés au milieu de la saison (ex: janvier et février pour l'hiver).
  3. Le mois de fin avec une condition restrictive sur le jour (ex: avant le 20 mars).
* Logique d'exclusion : L'utilisation des elif permet de s'assurer qu'une date ne peut appartenir qu'à une seule saison, en suivant l'ordre chronologique de l'année.


```python
m = int(input("Mois (1-12) :"))
j = int(input("Jour (1-31):"))

if ( m == 12 and j >= 22 ) or ( m <= 2 ) or ( m == 3 and j < 20):
  saison = "Hiver"
elif ( m == 3 and j >= 20) or ( m <= 5 ) or ( m == 6 and j < 21):
    saison = "Printemps"
elif ( m == 6 and j >= 21 ) or ( m <= 8) or ( m == 9 and j < 23):
    saison = "Été"
elif ( m == 9 and j >= 23 ) or ( m <= 11) or ( m == 12 and j < 22):
    saison = "Automne"
else:
    saison = "Date invalide"

print(saison)
```

##### Points clés de l'exercice :

* Rigueur logique : Cet exercice montre l'importance de bien parenthéser les conditions and et or. Une petite erreur sur un signe (comme < au lieu de <=) fausserait les dates charnières (équinoxes et solstices).
* Gestion des cycles : Le cas de l'hiver est particulier car il commence en fin d'année (m=12) et se poursuit au début de la suivante (m=1,2), ce qui demande une attention particulière dans les tests logiques.
* Efficacité : La structure en if/elif est idéale ici car elle traite les périodes de l'année de manière séquentielle et s'arrête dès que la bonne saison est trouvée.


#### Exercice 8 : Génération d'un intervalle croissant

##### L'objectif:

Créer un programme qui liste tous les entiers compris entre une borne de départ et une borne de fin. C'est l'exercice parfait pour maîtriser la fonction range() et comprendre comment Python gère les limites d'une séquence.

##### Ma stratégie de résolution :

* Utilisation de range() : En Python, la fonction range(début, fin) s'arrête juste avant la valeur de fin. Pour afficher aussi vfin, j'ai ajouté +1 à la borne supérieure.
* Gestion de l'affichage : Par défaut, print() revient à la ligne à chaque fois. Pour obtenir un affichage horizontal comme dans le jeu de tests, j'utilise l'argument end=" ".
* Comportement automatique : Si l'utilisateur saisit une valeur de départ supérieure à la valeur de fin (ex: 5 vers 0), la boucle ne s'exécutera tout simplement pas, respectant ainsi le silence demandé par le jeu de tests.

```python
vdep = int(input("Entrer une valeur de départ :"))
vfin = int(input("Entrer une valeur de fin :"))

for i in range(vdep,vfin+1):
    print(i, end=" ")
```

##### Points clés de l'exercice :

* Inclusivité : Le fait que range soit exclusif sur sa borne de fin est une source d'erreur classique ; l'astuce du +1 est donc indispensable ici.
* Flexibilité du print : L'argument end est très pratique pour personnaliser la mise en forme des sorties et rendre les résultats plus lisibles.
* Logique de boucle : Cet exercice montre que Python gère nativement le cas où l'intervalle est impossible (départ > fin) en ne retournant rien, ce qui évite de devoir ajouter une condition if supplémentaire.

#### Exercice 9 : Tables de multiplication (Boucles imbriquées)

##### L'objectif :

L'idée est de générer une table les tables de multiplication de 1 à 10) sous forme de grille. C'est l'exercice parfait pour apprendre à parcourir deux dimensions (lignes et colonnes) de manière automatisée.

##### Ma stratégie de résolution :

* Double itération : J'utilise deux boucles for. La première gère les lignes (la table de 1, puis de 2, etc.) et la seconde gère chaque multiplication à l'intérieur de cette ligne.
* Formatage de l'espace : Pour que les nombres ne soient pas collés et que les colonnes soient bien droites (que le chiffre 4 soit bien aligné sous le 12 par exemple), j'utilise le formatage :4d. Cela réserve 4 espaces pour chaque nombre.
* Gestion des sauts de ligne : J'utilise end="" pour rester sur la même ligne pendant le calcul d'une table, puis un print() vide à la fin de la boucle interne pour passer à la ligne suivante avant de commencer la table suivante.

```python
for i in range(1,11):
    for j in range(1,11):
        print(f"{i * j:4d}", end="")
        print()
```

#### Points clés de l'exercice :

* Logique 2D : Comprendre l'imbrication des boucles est essentiel pour manipuler des structures complexes comme des matrices ou des grilles de données.
* Formatage d'affichage : L'astuce {value:4d} est très utile en ingénierie pour présenter des résultats numériques de manière lisible et organisée.
* Contrôle du flux : La distinction entre l'affichage partiel (avec end="") et le saut de ligne final (le print() vide) est la clé pour structurer n'importe quel affichage en tableau.

#### Exercice 10 : Recherche des diviseurs d'un nombre
##### L'objectif :

Le but de cet exercice est de trouver tous les nombres qui peuvent diviser un entier donné sans laisser de reste. On doit les afficher du plus grand au plus petit et compter combien il y en a au total.

##### Ma stratégie de résolution :

* La boucle inversée : Pour afficher les diviseurs en ordre décroissant, j'utilise range(nb, 0, -1). Cela commence au nombre lui-même et descend jusqu'à 1.
* Le test de division : Pour savoir si un nombre est un diviseur, j'utilise l'opérateur modulo %. Si nb % i == 0, cela veut dire que le reste est nul et que i est un diviseur.
* Le compteur : Je crée une variable c (initialisée à 0) que j'augmente de 1 chaque fois que je trouve un diviseur. À la fin, j'affiche cette valeur pour répondre à la consigne.

```python
nb = int(input("Entrer un entier :"))
c = 0
for i in range(nb, 0, -1):
    if nb % i == 0:
        print(i, end=" ")
        c += 1
print(" \n Le nombre total des diviseurs :",c)
```

##### Points clés de l'exercice :

* Maîtrise de range() : J'ai appris que l'on peut faire une boucle qui recule en utilisant un "pas" de -1. C'est très utile pour trier des données à l'envers.
* Utilité du modulo : L'opérateur % est l'outil principal en programmation pour vérifier la divisibilité entre deux nombres.
* Logique de comptage : Utiliser un compteur à l'intérieur d'une condition if permet de filtrer et de compter uniquement les éléments qui nous intéressent.
