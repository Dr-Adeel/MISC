# **Exerice 2 : Classes Bâtiment et Maison**

# **Q1 :  Classe Batiment**

```
class Bâtiment:
    def __init__(self, adresse=""):
        self.__adresse = adresse  # Attribut privé

    # Accesseur (getter)
    def get_adresse(self):
        return self.__adresse

    # Mutateur (setter)
    def set_adresse(self, adresse):
        self.__adresse = adresse

    # Méthode toString
    def __str__(self):
        return f"Bâtiment - Adresse: {self.__adresse}"
```

# **Q2 : Classe Maison (heritant de Batiment)**

```
class Maison(Bâtiment):
    def __init__(self, adresse="", nbPieces=0):
        super().__init__(adresse)  # Appel du constructeur de la classe parent
        self.__nbPieces = nbPieces  # Attribut privé

    # Accesseur (getter) pour nbPieces
    def get_nbPieces(self):
        return self.__nbPieces

    # Mutateur (setter) pour nbPieces
    def set_nbPieces(self, nbPieces):
        self.__nbPieces = nbPieces

    # Méthode toString redéfinie
    def __str__(self):
        return f"Maison - Adresse: {self.get_adresse()}, Nombre de pièces: {self.__nbPieces}"
```

# **Exercice 3 : Hiérarchie Personne-Employé-Chef-Directeur**

# **Q1 : Les classes Personne, Employé, Chef et Directeur**

```
class Personne:
    def __init__(self, nom="", prenom="", date_naissance=""):
        self.__nom = nom
        self.__prenom = prenom
        self.__date_naissance = date_naissance

    def afficher(self):
        print(f"{self.__prenom} {self.__nom}, Né le: {self.__date_naissance}")

class Employe(Personne):
    def __init__(self, nom="", prenom="", date_naissance="", salaire=0):
        super().__init__(nom, prenom, date_naissance)
        self.__salaire = salaire

    def afficher(self):
        super().afficher()
        print(f"Salaire: {self.__salaire}€")

class Chef(Employe):
    def __init__(self, nom="", prenom="", date_naissance="", salaire=0, service=""):
        super().__init__(nom, prenom, date_naissance, salaire)
        self.__service = service

    def afficher(self):
        super().afficher()
        print(f"Service: {self.__service}")

class Directeur(Chef):
    def __init__(self, nom="", prenom="", date_naissance="", salaire=0, service="", societe=""):
        super().__init__(nom, prenom, date_naissance, salaire, service)
        self.__societe = societe

    def afficher(self):
        super().afficher()
        print(f"Société: {self.__societe}")
```

# **Exercice 4 : Héritage et polymorphisme pour le calcul de salaire des employés**

# **Q1. Classe Employé**

```
from abc import ABC, abstractmethod

class Employe(ABC):
    def __init__(self, matricule, nom, prenom, date_naissance):
        self.__matricule = matricule
        self.__nom = nom
        self.__prenom = prenom
        self.__date_naissance = date_naissance

    @abstractmethod
    def get_salaire(self):
        pass

    def __str__(self):
        return f"{self.__prenom} {self.__nom}"
```

# **Q2 : Classes Ouvrier, Cadre, Patron**

```
class Ouvrier(Employe):
    SMIG = 2500

    def __init__(self, matricule, nom, prenom, date_naissance, date_entree):
        super().__init__(matricule, nom, prenom, date_naissance)
        self.__date_entree = date_entree

    def get_salaire(self):
        anciennete = 2024 - int(self.__date_entree[-4:])  # Calcule ancienneté
        salaire = Ouvrier.SMIG + (anciennete * 100)
        return min(salaire, Ouvrier.SMIG * 2)

    def __str__(self):
        return f"Ouvrier: {super().__str__()}, Salaire: {self.get_salaire()}$"

class Cadre(Employe):
    def __init__(self, matricule, nom, prenom, date_naissance, indice):
        super().__init__(matricule, nom, prenom, date_naissance)
        self.__indice = indice

    def get_salaire(self):
        salaires = {1: 13000, 2: 15000, 3: 17000, 4: 20000}
        return salaires.get(self.__indice, 0)

    def __str__(self):
        return f"Cadre: {super().__str__()}, Indice: {self.__indice}, Salaire: {self.get_salaire()}$"

class Patron(Employe):
    CA = 1000000  # Chiffre d'affaire commun

    def __init__(self, matricule, nom, prenom, date_naissance, pourcentage):
        super().__init__(matricule, nom, prenom, date_naissance)
        self.__pourcentage = pourcentage

    def get_salaire(self):
        return (Patron.CA * self.__pourcentage) / 100

    def __str__(self):
        return f"Patron: {super().__str__()}, Salaire annuel: {self.get_salaire()}$"
```

# **Q3 : Programme de test**

```
# Test
if __name__ == "__main__":
    # Création des employés
    e1 = Ouvrier("O001", "Dupont", "Jean", "1990", "2015")
    e2 = Ouvrier("O002", "Martin", "Marie", "1985", "2020")
    e3 = Cadre("C001", "Bernard", "Pierre", "1975", 1)
    e4 = Cadre("C002", "Thomas", "Julie", "1980", 3)
    e5 = Patron("P001", "Durand", "Marc", "1965", 5)

    # Affichage
    for emp in [e1, e2, e3, e4, e5]:
        print(emp)

    # Test CA patron
    print(f"\nCA patron: {Patron.CA}$")
```
