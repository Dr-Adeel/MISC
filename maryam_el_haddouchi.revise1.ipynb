```python
age = int(input("Entrez l'âge : "))

if age <= 5:
    print("OUPS pas suffisamment grand")
elif 6 <= age <= 9:
    print("Poussin ou Pupille")
elif 10 <= age <= 15:
    print("Minime ou Cadet")
elif 16 <= age <= 17:
    print("Junior")
else:
    print("Senior")
```

```python
jour = int(input("Jour (0=dimanche à 6=samedi) : "))
heure = int(input("Heure (0 à 24) : "))

ouvert = False

if 0 <= jour <= 6 and 0 <= heure <= 24:
    if jour == 0:
        ouvert = False
    elif jour == 6 and heure >= 14:
        ouvert = False
    elif 8 <= heure < 13 or 14 <= heure < 17:
        ouvert = True

print("Vrai" if ouvert else "Faux")
```

```python
n = int(input("Entrez un entier positif : "))
compteur = 0

for i in range(n, 0, -1):
    if n % i == 0:
        print(i, end=" ")
        compteur += 1

print("==>", compteur)
```
