### *Calcule* du périmètre et de l'aire d'un disque de r=4.56 cm de diamètre.

```python
import math

d=4.56
r = d/2
aire = math.pi * r**2

perimetre = math.pi * d

print ("le périmetre est ", round(perimetre, 2))
print ("L'aire est ", round(aire, 2))
```

### Calcul de la somme des fractions sous forme irréductible (sans le module `fractions`)

Nous allons d'abord définir des fonctions utilitaires pour simplifier une fraction et pour additionner deux fractions. La simplification utilisera la fonction `math.gcd` (plus grand commun diviseur).

```python
import math

def simplification_fraction(numerator, denominator):
    """Simplifie une fraction donnée à sa forme irréductible."""
    common_divisor = math.gcd(numerator, denominator)
    return numerator // common_divisor, denominator // common_divisor

def somme_fractions(n1, d1, n2, d2):
    """Additionne deux fractions (n1/d1 + n2/d2) et retourne le résultat sous forme irréductible."""
    # Trouver un dénominateur commun simple (produit des dénominateurs)
    deno_commun = d1 * d2

    # Ajuster les numérateurs
    new_n1 = n1 * d2
    new_n2 = n2 * d1

    # Somme des numérateurs
    sum_n = new_n1 + new_n2

    # Simplifier la fraction résultante
    return simplification_fraction(sum_n, deno_commun)

```

```python
# Fractions à additionner
f1_n, f1_d = 217, 440
f2_n, f2_d = 101, 256
f3_n, f3_d = 86, 71

print(f"Fractions à additionner : {f1_n}/{f1_d}, {f2_n}/{f2_d}, {f3_n}/{f3_d}")

# Étape 1 : Additionner les deux premières fractions
result_n12, result_d12 = somme_fractions(f1_n, f1_d, f2_n, f2_d)
print(f"\nRésultat de ({f1_n}/{f1_d}) + ({f2_n}/{f2_d}) = {result_n12}/{result_d12}")

# Étape 2 : Additionner le résultat de l'étape 1 avec la troisième fraction
final_n, final_d = somme_fractions(result_n12, result_d12, f3_n, f3_d)

print(f"\nLa somme finale irréductible est : {final_n}/{final_d}")
```

### Vérification du résultat avec le module `fractions`

Maintenant, nous utilisons le module `fractions` pour vérifier que notre calcul manuel est correct.

```python
from fractions import Fraction

# Effectuer la somme en utilisant le module fractions
fraction_sum_verify = Fraction(f1_n, f1_d) + Fraction(f2_n, f2_d) + Fraction(f3_n, f3_d)

print(f"Somme vérifiée avec le module fractions : {fraction_sum_verify}")

# Comparer les résultats
if final_n == fraction_sum_verify.numerator and final_d == fraction_sum_verify.denominator:
    print("Le résultat manuel correspond au résultat du module fractions. \u2705")
else:
    print("ATTENTION : Les résultats ne correspondent pas. \u274c")
```

### Trouver la plus grande de trois valeurs entières

Ce code prend trois entiers en entrée, puis détermine et affiche la plus grande de ces valeurs.

```python
# Jeux de tests :
# 1 6 15 ==> 15
# 15 6 1 ==> 15
# 10 20 -5 ==> 20
# 0 0 0 ==> 0
# 3 -9 123 ==> 123

# Saisie des trois valeurs entières
print("Veuillez entrer trois nombres entiers.")
valeur1 = int(input("Entrez le premier entier : "))
valeur2 = int(input("Entrez le deuxième entier : "))
valeur3 = int(input("Entrez le troisième entier : "))

# Déterminer la plus grande valeur
plus_grand = max(valeur1, valeur2, valeur3)

# Afficher le résultat
print(f"La plus grande des trois valeurs est : {plus_grand}")
```

### Détermination de la catégorie sportive par âge

Ce code vous demandera d'entrer l'âge d'un enfant et affichera sa catégorie sportive correspondante.

```python
# Jeux de tests :
# 7 ==> Poussin ou Pupille
# 10 ==> Minime ou Cadet
# 12 ==> Minime ou Cadet
# 16 ==> Junior
# 18 ==> Senior
# 20 ==> Senior
# 5 ==> OUPS pas suffisamment grand

# Saisie de l'âge
age = int(input("Veuillez entrer l'âge de l'enfant (un entier positif) : "))

# Détermination de la catégorie
if age <= 5:
    categorie = "OUPS pas suffisamment grand"
elif age <= 9: # 6 à 9 ans
    categorie = "Poussin ou Pupille"
elif age <= 15: # 10 à 15 ans (10-11 ou après 12 jusqu'à 15 avant Junior)
    categorie = "Minime ou Cadet"
elif age <= 17: # 16 à 17 ans (après 16 inclus, avant Senior)
    categorie = "Junior"
else: # 18 ans et plus
    categorie = "Senior"

# Affichage du résultat
print(f"Pour un âge de {age} ans, la catégorie sportive est : {categorie}")
```

### Vérification de l'ouverture d'un guichet

Ce code détermine si un guichet est ouvert à un jour et une heure donnés, en tenant compte des horaires d'ouverture et des exceptions.

```python
# Jeux de tests :
# 1 9 ==> Vrai
# 0 5 ==> Faux
# 6 15 ==> Faux
# 6 13 ==> Vrai
# 3 16 ==> Vrai

# Saisie du numéro du jour de la semaine et de l'heure
jour_semaine = int(input("Veuillez entrer le numéro du jour de la semaine (0=Dimanche, ..., 6=Samedi) : "))
heure_journee = int(input("Veuillez entrer l'heure de la journée (entre 0 et 24) : "))

# Initialisation de la variable d'état du guichet
est_ouvert = False

# Détermination de l'état du guichet
if jour_semaine == 0: # Dimanche
    est_ouvert = False
elif jour_semaine == 6: # Samedi
    # Ouvert le samedi matin (8h à 13h inclus)
    if heure_journee >= 8 and heure_journee <= 13:
        est_ouvert = True
    else:
        est_ouvert = False # Fermé l'après-midi et en dehors des heures du matin
elif 1 <= jour_semaine <= 5: # Lundi au Vendredi
    # Ouvert de 8h à 13h inclus OU de 14h à 17h inclus
    if (heure_journee >= 8 and heure_journee <= 13) or \
       (heure_journee >= 14 and heure_journee <= 17):
        est_ouvert = True
    else:
        est_ouvert = False
# Pour les jours non valides (e.g., hors 0-6), est_ouvert reste False (par défaut ou implicitement)

# Affichage du résultat
print(est_ouvert)
```

### Détermination de la saison par date

Ce code prend un numéro de mois (1-12) et un jour du mois (1-31), puis détermine la saison correspondante (hiver, printemps, été, automne) en fonction des dates de transition spécifiées.

```python
# Jeux de tests :
# 12 31 ==> hiver
# 3 19 ==> hiver
# 3 20 ==> printemps
# 5 31 ==> printemps
# 6 20 ==> printemps
# 6 21 ==> ete
# 8 10 ==> ete
# 9 22 ==> ete
# 9 23 ==> automne
# 10 30 ==> automne
# 12 21 ==> automne
# 12 22 ==> hiver

# Saisie du mois et du jour
m = int(input("Veuillez entrer le numéro du mois (1 à 12) : "))
j = int(input("Veuillez entrer le jour du mois (1 à 31) : "))

# Détermination de la saison
if (m == 12 and j >= 22) or (m == 1) or (m == 2) or (m == 3 and j < 20):
    saison = "hiver"
elif (m == 3 and j >= 20) or (m == 4) or (m == 5) or (m == 6 and j < 21):
    saison = "printemps"
elif (m == 6 and j >= 21) or (m == 7) or (m == 8) or (m == 9 and j < 23):
    saison = "ete"
elif (m == 9 and j >= 23) or (m == 10) or (m == 11) or (m == 12 and j < 22):
    saison = "automne"
else:
    saison = "Date invalide ou non couverte par les règles."

# Affichage du résultat
print(f"Le {j}/{m} correspond à la saison : {saison}")
```

### Affichage des entiers d'un intervalle en ordre croissant

Ce code prend une valeur de départ et une valeur de fin pour un intervalle, puis affiche tous les entiers compris dans cet intervalle en ordre croissant. Si la valeur de départ est supérieure à la valeur de fin, aucun entier ne sera affiché.

```python
# Jeux de tests :
# -3 5 ==> -3 -2 -1 0 1 2 3 4 5
# 0 1 ==> 0 1
# 5 0 ==>

# Saisie des valeurs de départ et de fin
vdep = int(input("Veuillez entrer la valeur de départ (entier) : "))
vfin = int(input("Veuillez entrer la valeur de fin (entier) : "))

# Vérification de l'ordre de l'intervalle et affichage
if vdep <= vfin:
    for i in range(vdep, vfin + 1):
        print(i, end=" ")
    print() # Pour passer à la ligne après l'affichage
# Si vdep > vfin, rien n'est affiché (comportement par défaut)
```

### Trouver les diviseurs d'un entier positif en ordre décroissant

Ce code prend un entier positif en entrée, puis calcule et affiche tous ses diviseurs en ordre décroissant. Le dernier nombre affiché sera le compte total des diviseurs.

```python
# Jeux de tests :
# 24 ==> 24 12 8 6 4 3 2 1 ==> 8
# 15 ==> 15 5 3 1 ==> 4
# 123 ==> 123 41 3 1 ==> 4
# 423 ==> 423 141 47 9 3 1 ==> 6

# Saisie de l'entier positif
nombre = int(input("Veuillez entrer un entier positif : "))

# Liste pour stocker les diviseurs
diviseurs = []

# Trouver les diviseurs en ordre décroissant
# On peut itérer de 1 jusqu'à la racine carrée du nombre pour optimiser
# Mais pour afficher en ordre décroissant directement, une approche simple est de parcourir de 'nombre' jusqu'à 1.
for i in range(1, nombre + 1):
    if nombre % i == 0:
        diviseurs.append(i)

# Trier les diviseurs en ordre décroissant
diviseurs.sort(reverse=True)

# Afficher les diviseurs
for diviseur in diviseurs:
    print(diviseur, end=" ")

# Afficher le nombre total de diviseurs
print(f"==> {len(diviseurs)}")
```
