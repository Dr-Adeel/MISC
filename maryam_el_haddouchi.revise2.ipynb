```python
def factoriel(n):
    resultat = 1
    for i in range(1, n + 1):
        resultat *= i
    return resultat
```

```python
n = int(input("Entrez un entier positif : "))
for j in range(n + 1):
    print(f"{j}! = {factoriel(j)}")
```

```python
def icarre(n):
    somme = 0
    impair = 1

    for _ in range(n):
        somme += impair
        impair += 2   

    return somme
```

```python
def mult0(a, b):
    resultat = 0
    for _ in range(b):
        resultat += a
    return resultat
```

```python
def valeur_absolue(x):
    if x < 0:
        return -x
    return x

def signe(x):
    if x < 0:
        return -1
    return 1

def mult(a, b):
   
    produit = mult0(valeur_absolue(a), valeur_absolue(b))
    
    
    s = signe(a) * signe(b)
    
    return s * produit
```

```python
a = int(input("Entrez a : "))
b = int(input("Entrez b : "))

print("mult0 =", mult0(valeur_absolue(a), valeur_absolue(b)))
print("mult =", mult(a, b))
```

```python
def noccurs(lt, e):
    if not isinstance(lt, list):
        return -1
    
    compteur = 0
    for x in lt:
        if x == e:
            compteur += 1
    return compteur
```

```python
def traceMT(m):
    n = len(m)
    trace = 0
    for i in range(n):
        trace += m[i][i]
    return trace
```
