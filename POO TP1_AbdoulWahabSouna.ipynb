```python
from math import gcd
from fractions import Fraction
n1, d1 = 217, 440
n2, d2 = 101, 256
n3, d3 = 86, 71
den = d1 * d2 * d3
num = n1 * d2 * d3 + n2 * d1 * d3 + n3 * d1 * d2
g = gcd(num, den)
num_red = num // g
den_red = den // g
print("Somme irréductible (sans fractions) :", num_red, "/", den_red)
f = Fraction(n1, d1) + Fraction(n2, d2) + Fraction(n3, d3)
print("Vérification avec fractions :", f)
```

```python
n = int(input("Entrez un entier positif : "))
if n <= 0:
    print("Le nombre doit être positif !")
else:
    diviseurs = []  # Liste pour stocker les diviseurs
    compteur = 0    # Compteur de diviseurs
    for i in range(1, n + 1):
        if n % i == 0:  # i est un diviseur de n
            diviseurs.append(i)
            compteur += 1
    print(f"Diviseurs de {n} en ordre décroissant :")
    for diviseur in reversed(diviseurs):
        print(diviseur, end=" ")
    print(compteur)
```

```python
n = int(input("Entrez un entier positif : "))
diviseurs = []
for i in range(1, n + 1):
    if n % i == 0:
        diviseurs.append(i)
diviseurs.sort(reverse=True)
for d in diviseurs:
    print(d, end=" ")
print("=>", len(diviseurs))
```
