```python
#  PARC AUTOMOBILE 

class Vehicule:
    __compteur_matricule = 0
    
    def __init__(self, annee, prix):
        Vehicule.__compteur_matricule += 1
        self.__matricule = Vehicule.__compteur_matricule
        self.__annee = annee
        self.__prix = prix
    
    @property
    def matricule(self):
        return self.__matricule
    
    @property
    def annee(self):
        return self.__annee
    
    @annee.setter
    def annee(self, valeur):
        self.__annee = valeur
    
    @property
    def prix(self):
        return self.__prix
    
    @prix.setter
    def prix(self, valeur):
        self.__prix = valeur
    
    def demarrer(self):
        pass
    
    def accelerer(self):
        pass
    
    def __str__(self):
        return f"Matricule: {self.__matricule}, Année: {self.__annee}, Prix: {self.__prix} €"
    
    def __repr__(self):
        return self.__str__()


class Voiture(Vehicule):
    def __init__(self, annee, prix):
        super().__init__(annee, prix)
    
    def demarrer(self):
        return "La voiture démarre avec un démarreur électronique"
    
    def accelerer(self):
        return "La voiture accélère rapidement et silencieusement"


class Camion(Vehicule):
    def __init__(self, annee, prix, capacite_charge=0):
        super().__init__(annee, prix)
        self.__capacite_charge = capacite_charge
    
    @property
    def capacite_charge(self):
        return self.__capacite_charge
    
    @capacite_charge.setter
    def capacite_charge(self, valeur):
        self.__capacite_charge = valeur
    
    def demarrer(self):
        return "Le camion démarre avec un gros moteur diesel"
    
    def accelerer(self):
        return "Le camion accélère lentement mais avec puissance"
    
    def __str__(self):
        base = super().__str__()
        return f"{base}, Capacité de charge: {self.__capacite_charge} kg"



class TestVehicule:
    @staticmethod
    def tester():
        
        print("\n1. Création de véhicules:")
        
        voiture1 = Voiture(2022, 25000)
        print(f"Voiture créée: {voiture1}")
        
        camion1 = Camion(2020, 75000, 15000)
        print(f"Camion créé: {camion1}")
        
        voiture2 = Voiture(2023, 32000)
        print(f"Voiture créée: {voiture2}")
        
        print("\n2. Test des méthodes:")
        print(f"Voiture - {voiture1.demarrer()}")
        print(f"Voiture - {voiture1.accelerer()}")
        
        print(f"\nCamion - {camion1.demarrer()}")
        print(f"Camion - {camion1.accelerer()}")
        
        print("\n3. Test des accesseurs et mutateurs:")
        
        voiture1.annee = 2023
        voiture1.prix = 27000
        print(f"Voiture modifiée: {voiture1}")
        
        camion1.capacite_charge = 18000
        print(f"Camion modifié: {camion1}")
        
        print("\n4. Matricules des véhicules:")
        print(f"Matricule voiture1: {voiture1.matricule}")
        print(f"Matricule camion1: {camion1.matricule}")
        print(f"Matricule voiture2: {voiture2.matricule}")
        





```

```python
if __name__ == "__main__":
    # Exécution des tests
    TestVehicule.tester()
    
    # Exemple supplémentaire: création de plusieurs véhicules
    print("\nEXEMPLE SUPPLÉMENTAIRE: CRÉATION MULTIPLE")
    
    
    vehicules = [
        Voiture(2021, 20000),
        Camion(2019, 60000, 12000),
        Voiture(2023, 35000),
        Camion(2022, 80000, 20000)
    ]
    
    for i, vehicule in enumerate(vehicules, 1):
        print(f"\nVéhicule {i}: {vehicule}")
        print(f"  - Démarrage: {vehicule.demarrer()}")
        print(f"  - Accélération: {vehicule.accelerer()}")
```

```python
#  Classes Batiment et Maison


# Classe Batiment
class Batiment:
    
    def __init__(self, adresse=None):
        if adresse is None:
            self.__adresse = "Adresse non spécifiée"
        else:
            self.__adresse = adresse
    
   
    def get_adresse(self):
        return self.__adresse
    

    def set_adresse(self, nouvelle_adresse):
        self.__adresse = nouvelle_adresse
    
   
    def __str__(self):
        return f"Batiment[Adresse: {self.__adresse}]"


# Classe Maison qui hérite de Batiment
class Maison(Batiment):
    def __init__(self, adresse=None, nbPieces=0):
        # Appel du constructeur parent
        super().__init__(adresse)
        self.__nbPieces = nbPieces
    
    def get_nbPieces(self):
        return self.__nbPieces
    
    def set_nbPieces(self, nouveau_nb):
        self.__nbPieces = nouveau_nb
    
    def __str__(self):
        return f"Maison[Adresse: {self.get_adresse()}, Nombre de pièces: {self.__nbPieces}]"
```

```python
# Classes Personne, Employé, Chef, Directeur


# Classe Personne 
class Personne:
    def __init__(self, nom, prenom, date_naissance):
        self.__nom = nom
        self.__prenom = prenom
        self.__date_naissance = date_naissance
    
    def get_nom(self):
        return self.__nom
    
    def get_prenom(self):
        return self.__prenom
    
    def get_date_naissance(self):
        return self.__date_naissance
    
    def afficher(self):
        return f"Personne: {self.__prenom} {self.__nom}, Né(e) le: {self.__date_naissance}"


class Employe(Personne):
    def __init__(self, nom, prenom, date_naissance, salaire):
        super().__init__(nom, prenom, date_naissance)
        self.__salaire = salaire
    
    def get_salaire(self):
        return self.__salaire
    
    def afficher(self):
        base_info = super().afficher()
        return f"{base_info}, Salaire: {self.__salaire}€"


class Chef(Employe):
    def __init__(self, nom, prenom, date_naissance, salaire, service):
        super().__init__(nom, prenom, date_naissance, salaire)
        self.__service = service
    
    def get_service(self):
        return self.__service
    
    def afficher(self):
        base_info = super().afficher()
        return f"{base_info}, Service: {self.__service}"


class Directeur(Chef):
    def __init__(self, nom, prenom, date_naissance, salaire, service, societe):
        super().__init__(nom, prenom, date_naissance, salaire, service)
        self.__societe = societe
    
    def get_societe(self):
        return self.__societe
    
    def afficher(self):
        base_info = super().afficher()
        return f"{base_info}, Société: {self.__societe}"

```

```python


# Classe Employé de base
class Employe:
    def __init__(self, matricule, nom, prenom, date_naissance):
        self.__matricule = matricule
        self.__nom = nom
        self.__prenom = prenom
        self.__date_naissance = date_naissance
    
    def get_matricule(self):
        return self.__matricule
    
    def get_nom(self):
        return self.__nom
    
    def get_prenom(self):
        return self.__prenom
    
    def get_date_naissance(self):
        return self.__date_naissance
    
    def get_salaire(self):
        return 0  
    
    def __str__(self):
        return f"{self.__matricule} - {self.__prenom} {self.__nom}"


# Classe Ouvrier
class Ouvrier(Employe):
    SMIG = 2500  
    
    def __init__(self, matricule, nom, prenom, date_naissance, date_entree):
        Employe.__init__(self, matricule, nom, prenom, date_naissance)
        self.__date_entree = date_entree
    
    def get_salaire(self):
        anciennete = 2024 - self.__date_entree
        salaire = Ouvrier.SMIG + (anciennete * 100)
        
        if salaire > Ouvrier.SMIG * 2:
            salaire = Ouvrier.SMIG * 2
        
        return salaire
    
    def __str__(self):
        return f"Ouvrier: {super().__str__()}, Entré en {self.__date_entree}, Salaire: {self.get_salaire()}€"


# Classe Cadre
class Cadre(Employe):
    def __init__(self, matricule, nom, prenom, date_naissance, indice):
        Employe.__init__(self, matricule, nom, prenom, date_naissance)
        self.__indice = indice
    
    def get_salaire(self):
        if self.__indice == 1:
            return 13000
        elif self.__indice == 2:
            return 15000
        elif self.__indice == 3:
            return 17000
        elif self.__indice == 4:
            return 20000
        else:
            return 0
    
    def __str__(self):
        return f"Cadre: {super().__str__()}, Indice {self.__indice}, Salaire: {self.get_salaire()}€"


# Classe Patron
class Patron(Employe):
    CA = 1000000  
    
    def __init__(self, matricule, nom, prenom, date_naissance, pourcentage):
        Employe.__init__(self, matricule, nom, prenom, date_naissance)
        self.__pourcentage = pourcentage
    
    def get_salaire(self):
        return (Patron.CA * self.__pourcentage) / 100
    
    def __str__(self):
        salaire_annuel = self.get_salaire()
        salaire_mensuel = salaire_annuel / 12
        return f"Patron: {super().__str__()}, {self.__pourcentage}% du CA, Salaire: {salaire_mensuel:.0f}€/mois"

```

```python

#  test 
def main():


    
    # Ouvriers
    o1 = Ouvrier("EMP001", "Martin", "Pierre", "15/05/1985", 2015)
    o2 = Ouvrier("EMP002", "Dubois", "Julie", "22/08/1990", 2022)
    
    # Cadres
    c1 = Cadre("EMP101", "Durand", "Sophie", "12/06/1980", 2)
    c2 = Cadre("EMP102", "Leroy", "Thomas", "30/11/1975", 4)
    
    # Patrons
    p1 = Patron("DIR001", "Bernard", "Robert", "15/02/1960", 10)
    
    print("Ouvriers:")
    print(f"  - {o1}")
    print(f"  - {o2}")
    
    print("\nCadres:")
    print(f"  - {c1}")
    print(f"  - {c2}")
    
    print("\nPatrons:")
    print(f"  - {p1}")

    print("2. Comparaison des salaires:")
    
    employes = [o1, o2, c1, c2, p1]
    
    print("\nListe des salaires mensuels:")
    for emp in employes:
        salaire = emp.get_salaire()
        if isinstance(emp, Patron):
            salaire_mensuel = salaire / 12
            print(f"  {emp.get_prenom()} {emp.get_nom()}: {salaire_mensuel:.0f}€/mois (patron)")
        else:
            print(f"  {emp.get_prenom()} {emp.get_nom()}: {salaire}€/mois")
    
    total = 0
    for emp in employes:
        if isinstance(emp, Patron):
            total += emp.get_salaire() / 12
        else:
            total += emp.get_salaire()
    
    print(f"\nMasse salariale mensuelle totale: {total:.0f}€")
    
 
    


if __name__ == "__main__":
    main()
```
