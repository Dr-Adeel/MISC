**Exercice 1**

```python
def factoriel(n):
  liste = []
  r = 1
  if n==0:
    liste.append(1)
  for i in range(1,n+1):
    r *= i
    liste.append(r)
  return liste
```

```python
factoriel(4)
```

```python
print(len(str(factoriel(62))))
```

**Exercice 2**

```python
def icarre(n):
  s = 0
  for i in range(n):
    imp = 2*i+1
    s += imp
  return s
```

**Exercice 3**

Question a

```python
def mult0(a,b):
  s = 0
  for i in range(b):
    s += a
  return s
```

Question b

```python
def signe(x):
  if x>0:
    return 1
  elif x<0:
    return -1
  else:
    return 0
```

```python
def val_abs(x):
  if x<0:
    return -x
  else:
    return x
```

```python
def mult(a,b):
  s = signe(a)*signe(b)
  if s < 0:
    return -mult0(val_abs(a),val_abs(b))
  else:
    return mult0(val_abs(a),val_abs(b))
```

Question c

```python
print(mult(10,5))
mult(-21,-123)
```

**Exercice 5**

```python
def noccurs(lt, e):
    
    if not isinstance(lt, list):
        return -1

    compteur = 0
    for element in lt:
        if element == e:
            compteur += 1
    return compteur


print("\nTests de la fonction noccurs:")
print(f'noccurs(["a",1,2,3,2,2,5,2], 2) = {noccurs(["a",1,2,3,2,2,5,2], 2)}')
print(f'noccurs(2, 2) = {noccurs(2, 2)}')
print(f'noccurs(["a",1,2,3,2,2,5,2], 9) = {noccurs(["a",1,2,3,2,2,5,2], 9)}')
```

**Exercice 7**

```python
def traceMT(m):
    n = len(m)
    s = 0
    for i in range(n):
        s += m[i][i]
    return s
```
