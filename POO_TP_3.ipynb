### Q1. Création de la classe `Véhicule`.

Here's the implementation for the `Vehicule` class. It's an abstract base class with a matricule counter, private attributes, and abstract methods `demarrer()` and `accélérer()`.

```python
from abc import ABC, abstractmethod

class Vehicule(ABC):
    _matricule_counter = 0  # Class variable for auto-incrementing matricule

    def __init__(self, annee_modele, prix):
        Vehicule._matricule_counter += 1
        self._matricule = Vehicule._matricule_counter # Auto-incremented matricule
        self._annee_modele = annee_modele
        self._prix = prix

    # Property for matricule (read-only as it's auto-incremented)
    @property
    def matricule(self):
        return self._matricule

    # Property for annee_modele (getter and setter)
    @property
    def annee_modele(self):
        return self._annee_modele

    @annee_modele.setter
    def annee_modele(self, value):
        if isinstance(value, int) and value > 0:
            self._annee_modele = value
        else:
            raise ValueError("L'année du modèle doit être un entier positif.")

    # Property for prix (getter and setter)
    @property
    def prix(self):
        return self._prix

    @prix.setter
    def prix(self, value):
        if isinstance(value, (int, float)) and value >= 0:
            self._prix = value
        else:
            raise ValueError("Le prix doit être un nombre positif.")

    # Abstract methods to be implemented by derived classes
    @abstractmethod
    def demarrer(self):
        pass

    @abstractmethod
    def accelerer(self):
        pass

    # ToString() equivalent
    def __str__(self):
        return f"Véhicule [Matricule: {self._matricule}, Année du modèle: {self._annee_modele}, Prix: {self._prix:.2f}€]"
```

### Q1. Écriture de la classe `Bâtiment`.

```python
class Batiment:
    def __init__(self, adresse=None):
        self._adresse = adresse

    @property
    def adresse(self):
        return self._adresse

    @adresse.setter
    def adresse(self, value):
        if isinstance(value, str) or value is None:
            self._adresse = value
        else:
            raise ValueError("L'adresse doit être une chaîne de caractères.")

    def __str__(self):
        return f"Bâtiment [Adresse: {self.adresse if self.adresse else 'Non spécifiée'}]"
```

### Q2. Écriture de la classe `Maison` héritant de `Bâtiment`.

```python
class Maison(Batiment):
    def __init__(self, adresse=None, nb_pieces=None):
        super().__init__(adresse)
        self._nb_pieces = nb_pieces

    @property
    def nb_pieces(self):
        return self._nb_pieces

    @nb_pieces.setter
    def nb_pieces(self, value):
        if isinstance(value, int) and value > 0 or value is None:
            self._nb_pieces = value
        else:
            raise ValueError("Le nombre de pièces doit être un entier positif.")

    def __str__(self):
        parent_str = super().__str__()
        return f"Maison [{parent_str.split('[')[1][:-1]}, Nb Pièces: {self.nb_pieces if self.nb_pieces is not None else 'Non spécifié'}]"
```

### Exercice 4: Q2. Création des `Ouvrier`, `Cadre` et `Patron`.

```python
from datetime import date

class Ouvrier(Employe):
    SMIG = 2500 # Class variable for SMIG

    def __init__(self, nom, prenom, date_naissance, date_entree):
        super().__init__(nom, prenom, date_naissance)
        if isinstance(date_entree, str):
            self._date_entree = date.fromisoformat(date_entree)
        elif isinstance(date_entree, date):
            self._date_entree = date_entree
        else:
            raise ValueError("La date d'entrée doit être une chaîne (YYYY-MM-DD) ou un objet date.")

    @property
    def date_entree(self):
        return self._date_entree

    @date_entree.setter
    def date_entree(self, value):
        if isinstance(value, (str, date)):
            if isinstance(value, str):
                self._date_entree = date.fromisoformat(value)
            else:
                self._date_entree = value
        else:
            raise ValueError("La date d'entrée doit être une chaîne (YYYY-MM-DD) ou un objet date.")

    def GetSalaire(self):
        anciennete = date.today().year - self._date_entree.year
        salaire = Ouvrier.SMIG + anciennete * 100
        return min(salaire, Ouvrier.SMIG * 2)

    def __str__(self):
        return f"Ouvrier [Matricule: {self.matricule}, Nom: {self.nom}, Prénom: {self.prenom}, Date naissance: {self.date_naissance.isoformat()}, Date entrée: {self.date_entree.isoformat()}, Salaire: {self.GetSalaire():.2f}$]"


class Cadre(Employe):
    SALAIRES_INDICE = {1: 13000, 2: 15000, 3: 17000, 4: 20000}

    def __init__(self, nom, prenom, date_naissance, indice):
        super().__init__(nom, prenom, date_naissance)
        if indice not in Cadre.SALAIRES_INDICE:
            raise ValueError("L'indice doit être entre 1 et 4.")
        self._indice = indice

    @property
    def indice(self):
        return self._indice

    @indice.setter
    def indice(self, value):
        if value not in Cadre.SALAIRES_INDICE:
            raise ValueError("L'indice doit être entre 1 et 4.")
        self._indice = value

    def GetSalaire(self):
        return Cadre.SALAIRES_INDICE[self._indice]

    def __str__(self):
        return f"Cadre [Matricule: {self.matricule}, Nom: {self.nom}, Prénom: {self.prenom}, Date naissance: {self.date_naissance.isoformat()}, Indice: {self.indice}, Salaire: {self.GetSalaire():.2f}$]"


class Patron(Employe):
    CHIFFRE_AFFAIRE_COMMUN = 1000000 # Example common turnover

    def __init__(self, nom, prenom, date_naissance, pourcentage):
        super().__init__(nom, prenom, date_naissance)
        if not (0 <= pourcentage <= 100):
            raise ValueError("Le pourcentage doit être entre 0 et 100.")
        self._pourcentage = pourcentage

    @property
    def pourcentage(self):
        return self._pourcentage

    @pourcentage.setter
    def pourcentage(self, value):
        if not (0 <= value <= 100):
            raise ValueError("Le pourcentage doit être entre 0 et 100.")
        self._pourcentage = value

    def GetSalaire(self):
        return (Patron.CHIFFRE_AFFAIRE_COMMUN * self._pourcentage) / 100

    def __str__(self):
        return f"Patron [Matricule: {self.matricule}, Nom: {self.nom}, Prénom: {self.prenom}, Date naissance: {self.date_naissance.isoformat()}, Pourcentage: {self.pourcentage}%, Salaire annuel: {self.GetSalaire():.2f}$]"
```

### Exercice 4: Q1. Création de la classe `Employé`.

```python
from abc import ABC, abstractmethod
from datetime import date

class Employe(ABC):
    _matricule_counter = 0  # Class variable for auto-incrementing matricule

    def __init__(self, nom, prenom, date_naissance):
        Employe._matricule_counter += 1
        self._matricule = Employe._matricule_counter
        self._nom = nom
        self._prenom = prenom
        if isinstance(date_naissance, str):
            self._date_naissance = date.fromisoformat(date_naissance)
        elif isinstance(date_naissance, date):
            self._date_naissance = date_naissance
        else:
            raise ValueError("La date de naissance doit être une chaîne (YYYY-MM-DD) ou un objet date.")

    @property
    def matricule(self):
        return self._matricule

    @property
    def nom(self):
        return self._nom

    @nom.setter
    def nom(self, value):
        if isinstance(value, str) and len(value) > 0:
            self._nom = value
        else:
            raise ValueError("Le nom doit être une chaîne non vide.")

    @property
    def prenom(self):
        return self._prenom

    @prenom.setter
    def prenom(self, value):
        if isinstance(value, str) and len(value) > 0:
            self._prenom = value
        else:
            raise ValueError("Le prénom doit être une chaîne non vide.")

    @property
    def date_naissance(self):
        return self._date_naissance

    @date_naissance.setter
    def date_naissance(self, value):
        if isinstance(value, (str, date)):
            if isinstance(value, str):
                self._date_naissance = date.fromisoformat(value)
            else:
                self._date_naissance = value
        else:
            raise ValueError("La date de naissance doit être une chaîne (YYYY-MM-DD) ou un objet date.")

    def __str__(self):
        return f"Matricule: {self.matricule}, Nom: {self.nom}, Prénom: {self.prenom}, Date de naissance: {self.date_naissance.isoformat()}"

    @abstractmethod
    def GetSalaire(self):
        pass
```

### Exercice 3: Q1. Écriture des classes `Personne`, `Employé`, `Chef` et `Directeur`.

```python
from datetime import date

class Personne:
    def __init__(self, nom, prenom, date_naissance):
        self._nom = nom
        self._prenom = prenom
        # date_naissance can be a string or a date object
        if isinstance(date_naissance, str):
            self._date_naissance = date.fromisoformat(date_naissance) # Assuming 'YYYY-MM-DD' format
        elif isinstance(date_naissance, date):
            self._date_naissance = date_naissance
        else:
            raise ValueError("La date de naissance doit être une chaîne (YYYY-MM-DD) ou un objet date.")

    @property
    def nom(self):
        return self._nom

    @nom.setter
    def nom(self, value):
        if isinstance(value, str) and len(value) > 0:
            self._nom = value
        else:
            raise ValueError("Le nom doit être une chaîne non vide.")

    @property
    def prenom(self):
        return self._prenom

    @prenom.setter
    def prenom(self, value):
        if isinstance(value, str) and len(value) > 0:
            self._prenom = value
        else:
            raise ValueError("Le prénom doit être une chaîne non vide.")

    @property
    def date_naissance(self):
        return self._date_naissance

    @date_naissance.setter
    def date_naissance(self, value):
        if isinstance(value, (str, date)):
            if isinstance(value, str):
                self._date_naissance = date.fromisoformat(value)
            else:
                self._date_naissance = value
        else:
            raise ValueError("La date de naissance doit être une chaîne (YYYY-MM-DD) ou un objet date.")

    def Afficher(self):
        return f"Nom: {self._nom}, Prénom: {self._prenom}, Date de naissance: {self._date_naissance.isoformat()}"

class Employe(Personne):
    def __init__(self, nom, prenom, date_naissance, salaire):
        super().__init__(nom, prenom, date_naissance)
        self._salaire = salaire

    @property
    def salaire(self):
        return self._salaire

    @salaire.setter
    def salaire(self, value):
        if isinstance(value, (int, float)) and value >= 0:
            self._salaire = value
        else:
            raise ValueError("Le salaire doit être un nombre positif.")

    def Afficher(self):
        return f"{super().Afficher()}, Salaire: {self._salaire:.2f}€"

class Chef(Employe):
    def __init__(self, nom, prenom, date_naissance, salaire, service):
        super().__init__(nom, prenom, date_naissance, salaire)
        self._service = service

    @property
    def service(self):
        return self._service

    @service.setter
    def service(self, value):
        if isinstance(value, str) and len(value) > 0:
            self._service = value
        else:
            raise ValueError("Le service doit être une chaîne non vide.")

    def Afficher(self):
        return f"{super().Afficher()}, Service: {self._service}"

class Directeur(Chef):
    def __init__(self, nom, prenom, date_naissance, salaire, service, societe):
        super().__init__(nom, prenom, date_naissance, salaire, service)
        self._societe = societe

    @property
    def societe(self):
        return self._societe

    @societe.setter
    def societe(self, value):
        if isinstance(value, str) and len(value) > 0:
            self._societe = value
        else:
            raise ValueError("La société doit être une chaîne non vide.")

    def Afficher(self):
        return f"{super().Afficher()}, Société: {self._societe}"
```
