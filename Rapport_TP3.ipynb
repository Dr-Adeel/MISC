```python
#Question 2
class Voiture(Vehicule) :
    def __init__(self,modele,prix):
        super().__init__(modele,prix)
    def demarrer(self):
        print("La voiture demarre rapidement")
    def accelerer(self):
        print("Le voiture accelere rapidement")
class Camion(Vehicule) :
     def __init__(self,modele,prix):
        super().__init__(modele,prix)
     def demarrer(self):
         print("La camion demarre lentement")
     def accelerer(self):
         print("Le camion accelere lentement")
```

#### Exercice 2 : Bâtiments et Maisons (Constructeurs)

##### L'objectif :

Pratiquer l'héritage avec des classes représentant des constructions civiles. On se concentre sur l'initialisation des données et la représentation textuelle.

##### Ma stratégie de résolution :

* Constructeurs flexibles : J'ai utilisé des arguments par défaut (None) pour permettre de créer un bâtiment avec ou sans adresse.
* Extension : La classe Maison ajoute l'attribut nbPieces tout en réutilisant l'adresse de la classe Batiment.
* ToString : J'utilise super().toString() dans la classe Maison pour afficher l'adresse avant d'ajouter le nombre de pièces.

```python
#Question 1
class Batiment:
    def __init__(self,adresse=None):
        self.__adresse = adresse
    def getAdresse(self):
        return self.__adresse
    def setAdresse(self,adresse):
        self.__adresse = adresse
    def toString(self):
        return f"Adresse du batiment : {self.__adresse}"
```

```python
#Question 2
class Maison(Batiment):
    def __init__(self,adresse=None,nbPieces=None):
        super().__init__(adresse)
        self.__nbPieces = nbPieces
    def getNBpieces(self):
        return self.__nbPieces
    def setNBpieces(self,nbPieces):
        self.__nbPieces = nbPieces
    def toString(self):
        return f"{super().toString()} \n Nombre de pieces : {self.__nbPieces}"
```

```python
#Question 3
b1 = Batiment("9 Rue de l'Ésperance.")
m1 = Maison("Avenue la fayette",2)
print(b1.toString())
print(m1.toString())
```

#### Exercice 3 : Hiérarchie de l'entreprise (Polymorphisme)

##### L'objectif :

Créer une chaîne d'héritage longue (Personne → Employé → Chef → Directeur) et utiliser le polymorphisme pour l'affichage.

##### Ma stratégie de résolution :

* Cascade d'héritage : Chaque classe descend de la précédente et ajoute un champ spécifique (salaire, service ou société).
* Polymorphisme : La méthode Afficher() est redéfinie à chaque niveau. Elle appelle super().Afficher() pour montrer toutes les informations accumulées.
* Tableau d'objets : J'ai créé une liste contenant différents types d'objets. Une simple boucle suffit pour afficher tout le monde correctement.

```python
class Personne:
    def __init__(self,nom,prenom,date):
        self.__nom = nom
        self.__prenom = prenom
        self.__date = date
    def getNom(self):
        return self.__nom
    def getPrenom(self):
        return self.__prenom
    def getDate(self):
        return self.__date
    def setNom(self,nom):
        self.__nom = nom
    def setPrenom(self,prenom):
        self.__prenom = prenom
    def setDate(self,date):
        self.__date = date
    def Afficher(self):
        print(f"Info:\n Nom : {self.getNom()}")
        print(f"Prenom : {self.getPrenom()}")
        print(f"Date : {self.getDate()}")
```

```python
class Employe(Personne):
    def __init__(self,nom,prenom,date,salaire):
        super().__init__(nom,prenom,date)
        self.__salaire = salaire

    def getSalaire(self):
        return self.__salaire
    def setSalaire(self,salaire):
        self.__salaire = salaire

    def Afficher(self):
        super().Afficher()
        print(f"Salaire : {self.__salaire}")
```

```python
class Chef(Employe):
    def __init__(self,nom,prenom,date,salaire,service):
        super().__init__(nom,prenom,date,salaire)
        self.__service = service
    def getService(self):
        return self.__service
    def setService(self,service):
        self.__service = service
    def Afficher(self):
        super().Afficher()
        print(f"Service : {self.__service}")
```

```python
class Directeur(Chef):
    def __init__(self,nom,prenom,date,salaire,service,societe):
        super().__init__(nom,prenom,date,salaire,service)
        self.__societe = societe
    def getSociete(self):
        return self.__societe
    def setSociete(self,societe):
        self.__societe = societe
    def Afficher(self):
        super().Afficher()
        print(f"Societe : {self.__societe}")
```

```python
lista = []
lista.append(Employe("Boussigua","Amal","12/12/2006",6000000))
E2 = Employe("Yakdane","Mohammed","28/12/2002",1000000000)
lista.append(E2)
D1 = Directeur("Brataa","Othmane","05/09/2000",8000000,"IT","Aramex")
lista.append(D1)
C1 = Chef("Labbane","Salma","30/03/1999",900000,"IT")
lista.append(C1)
for i in lista:
    i.Afficher()
    print("*"*20)
```

##### Points clés de l'exercice :

* Liaison dynamique : Python sait exactement quelle version de Afficher() appeler selon le type de l'objet dans la liste.
* Structure en couches : L'héritage permet de spécialiser les classes de plus en plus.


#### Exercice 4 : Gestion avancée des salaires (Abstraction)

##### L'objectif :

Calculer les salaires de différents types d'employés avec des règles métiers précises (SMIG, ancienneté, indices, chiffre d'affaire).
+4

##### Ma stratégie de résolution :

* Méthode Abstraite : Dans la classe Employe, la méthode getSalaire() est définie mais vide, car chaque catégorie a son propre calcul.
* Logique Ouvrier : Le salaire est limité à deux fois le SMIG (5000 $) après avoir ajouté l'ancienneté.
* Logique Patron : Utilisation d'un pourcentage sur le chiffre d'affaire commun.

```python
#Question 1
import datetime
class Employe:
    def __init__(self,matricule,nom,prenom,date):
        self.__matricule = matricule
        self.__nom = nom
        self.__prenom = prenom
        self.__date = date
    def getSalaire(self):
        print(" ")
    def toString(self):
        return f"Matricule : {self.__matricule} \n Nom : {self.__nom} \n Prenom : {self.__prenom} \n Date : {self.__date}"

    def getMatricule(self):
        return self.__matricule
    def getNom(self):
        return self.__nom
    def getPrenom(self):
        return self.__prenom
    def getDate(self):
        return self.__date
    def setMatricule(self,matricule):
        self.__matricule = matricule
    def setNom(self,nom):
        self.__nom = nom
    def setPrenom(self,prenom):
        self.__prenom = prenom
    def setDate(self,date):
        self.__date = date
class Ouvrier(Employe):
    SMIG = 2500
    def __init__(self,matricule,nom,prenom,date,dateE):
        super().__init__(matricule,nom,prenom,date)
        self.__dateE = dateE

    def getSalaire(self):
        nbAnnee = datetime.datetime.now().year - self.__dateE
        salaire = Ouvrier.SMIG + (nbAnnee * 100)
        if salaire > self.SMIG * 2 :
            salaire = self.SMIG * 2
        return salaire
```

```python
#Question 2
class Cadre(Employe):
    def __init__(self,matricule,nom,prenom,date,indice):
        super().__init__(matricule,nom,prenom,date)
        self.__indice = indice

    def getSalaire(self):
        if self.__indice == 1 : return 13000
        elif self.__indice == 2 : return 15000
        elif self.__indice == 3 : return 17000
        elif self.__indice == 4 : return 20000
        else : return 0
class Patron(Employe):
    def __init__(self,matricule,nom,prenom,date,ca,pourcentage):
        super().__init__(matricule,nom,prenom,date)
        self.__ca = ca
        self.__pourcentage = pourcentage
    def getSalaire(self):
        salaire = self.__ca*(self.__pourcentage/100)
        return salaire
    def toString(self):
        return f"{super().toString()} \n Salaire annuel : {self.getSalaire()}"
```

```python
#Question 3
o1 = Ouvrier("1400","MOURAD","Miloud",1959,1991)
c1 = Cadre("2546","SMITH","Smith","1970",2)
p1 = Patron("1000","AZULUN","OZAI","2000",1000000,5)
print(o1.getSalaire())
print(c1.getSalaire())
print(p1.getSalaire())
print(o1.toString())
print(c1.toString())
print(p1.toString())
```

* Indépendance : Chaque classe fille (Ouvrier, Cadre, Patron) possède sa propre formule mathématique sans impacter les autres.
