## Exercice 1

```python
from abc import ABC, abstractmethod

# Un parc auto
class Vehicle(ABC):
    __compteur = 0  

    def __init__(self, annee, prix):
        Vehicle.__compteur += 1
        self.__matricule = Vehicle.__compteur
        self.__annee = annee
        self.__prix = prix

    def setAnnee(self, annee):
        self.__annee = annee

    def setPrix(self, prix):
        self.__prix = prix

    def getMatricule(self):
        return self.__matricule

    def getAnnee(self):
        return self.__annee

    def getPrix(self):
        return self.__prix

    @abstractmethod
    def demarrer(self):
        pass

    @abstractmethod
    def accelerer(self):
        pass

    def ToString(self):
        return "Mat : {0}\tAnnee : {1}\tPrix : {2}".format(
            self.__matricule, self.__annee, self.__prix
        )

    def __str__(self):
        return self.ToString()

class Voiture(Vehicle):
    def demarrer(self):
        print(f"Voiture {self.getMatricule()} démarre ")

    def accelerer(self):
        print(f"Voiture {self.getMatricule()} accélère ")


class Camion(Vehicle):
    def demarrer(self):
        print(f"Camion {self.getMatricule()} démarre ")

    def accelerer(self):
        print(f"Camion {self.getMatricule()} accélère lentement ")


# Test
if __name__ == "__main__":
    v1 = Voiture(2020, 15000)
    c1 = Camion(2018, 45000)
    v2 = Voiture(2022, 23000)

    print(v1)
    print(c1)
    print(v2)

    v1.demarrer()
    v1.accelerer()
```

## Exercice 2 : Sur les batiments

```python
class Batiment:
    def __init__(self, adresse=""):
        self.__adresse = adresse  # attribut privé

    def getAdresse(self):
        return self.__adresse

    def setAdresse(self, adresse):
        self.__adresse = adresse

    def ToString(self):
        return "Batiment : adresse = {0}".format(self.__adresse)

    def __str__(self):
        return self.ToString()


class Maison(Batiment):
    def __init__(self, adresse="", nbPieces=0):
        super().__init__(adresse)  # appel du constructeur Batiment
        self.__nbPieces = nbPieces  # attribut privé

    def getNbPieces(self):
        return self.__nbPieces

    def setNbPieces(self, nbPieces):
        self.__nbPieces = nbPieces

    def ToString(self):
        return "{0}\tNbPieces = {1}".format(super().ToString(), self.__nbPieces)

    def __str__(self):
        return self.ToString()


# --- Programme de test  ---
if __name__ == "__main__":
    b1 = Batiment()
    b2 = Batiment("10 Avenue de Paris")

    m1 = Maison()
    m2 = Maison("60 Rue Ferdinand Buisson, Calais", 5)

    print("=== Test Batiment ===")
    print(b1)
    print(b2)

    print("\n=== Test Maison ===")
    print(m1)
    print(m2)

    # Tests setters
    m1.setAdresse("EILCO, Calais")
    m1.setNbPieces(3)
    print("\nAprès modification m1 :", m1)
```

```python
class Personne:
    def __init__(self, nom, prenom, date_naissance):
        self.__nom = nom
        self.__prenom = prenom
        self.__date_naissance = date_naissance
    def getNom(self):
        return self.__nom

    def setNom(self, nom):
        self.__nom = nom

    def getPrenom(self):
        return self.__prenom

    def setPrenom(self, prenom):
        self.__prenom = prenom

    def getDateNaissance(self):
        return self.__date_naissance

    def setDateNaissance(self, date_naissance):
        self.__date_naissance = date_naissance

    def Afficher(self):
        print(f"Personne : Nom={self.__nom}, Prénom={self.__prenom}, DateNaissance={self.__date_naissance}")


class Employe(Personne):
    def __init__(self, nom, prenom, date_naissance, salaire):
        super().__init__(nom, prenom, date_naissance)
        self.__salaire = salaire

    def getSalaire(self):
        return self.__salaire

    def setSalaire(self, salaire):
        self.__salaire = salaire

    def Afficher(self):
        print(f"Employé : Nom={self.getNom()}, Prénom={self.getPrenom()}, "
              f"DateNaissance={self.getDateNaissance()}, Salaire={self.__salaire}")


class Chef(Employe):
    def __init__(self, nom, prenom, date_naissance, salaire, service):
        super().__init__(nom, prenom, date_naissance, salaire)
        self.__service = service

    def getService(self):
        return self.__service

    def setService(self, service):
        self.__service = service

    # Redéfinition Afficher
    def Afficher(self):
        print(f"Chef : Nom={self.getNom()}, Prénom={self.getPrenom()}, "
              f"DateNaissance={self.getDateNaissance()}, Salaire={self.getSalaire()}, "
              f"Service={self.__service}")


class Directeur(Chef):
    def __init__(self, nom, prenom, date_naissance, salaire, service, societe):
        super().__init__(nom, prenom, date_naissance, salaire, service)
        self.__societe = societe

    def getSociete(self):
        return self.__societe

    def setSociete(self, societe):
        self.__societe = societe

    def Afficher(self):
        print(f"Directeur : Nom={self.getNom()}, Prénom={self.getPrenom()}, "
              f"DateNaissance={self.getDateNaissance()}, Salaire={self.getSalaire()}, "
              f"Service={self.getService()}, Société={self.__societe}")
# -----------------------------
# Programme de test (Q2 + Q3)
# -----------------------------
if __name__ == "__main__":
    personnes = [
        Employe("Diallo", "Aminata", "1999-01-15", 1200),
        Employe("Traore", "Moussa", "1998-03-10", 1300),
        Employe("Benali", "Sara", "2000-07-22", 1250),
        Employe("Nguyen", "Minh", "1997-12-05", 1400),
        Employe("Kone", "Ibrahim", "1996-09-18", 1500),
        Chef("Dupont", "Karim", "1990-02-02", 2500, "Informatique"),
        Chef("Martin", "Nadia", "1989-06-30", 2600, "Finance"),
        Directeur("Bernard", "Yassine", "1985-11-12", 4000, "Direction", "SUNU Assurances")
    ]

    print("=== Affichage de toutes les personnes (polymorphisme) ===")
    for p in personnes:
        p.Afficher()
```

## Exercice 4

```python
from abc import ABC, abstractmethod
from datetime import date, datetime

class Employe(ABC):
    def __init__(self, matricule, nom, prenom, date_naissance):
        self.__matricule = matricule
        self.__nom = nom
        self.__prenom = prenom
        self.__date_naissance = date_naissance  # string "YYYY-MM-DD" ou date

    # --- Propriétés (get/set) ---
    def getMatricule(self):
        return self.__matricule

    def setMatricule(self, matricule):
        self.__matricule = matricule

    def getNom(self):
        return self.__nom

    def setNom(self, nom):
        self.__nom = nom

    def getPrenom(self):
        return self.__prenom

    def setPrenom(self, prenom):
        self.__prenom = prenom

    def getDateNaissance(self):
        return self.__date_naissance

    def setDateNaissance(self, date_naissance):
        self.__date_naissance = date_naissance

    # --- Méthode abstraite ---
    @abstractmethod
    def GetSalaire(self):
        pass

    # --- ToString ---
    def ToString(self):
        return "Matricule={0}\tNom={1}\tPrénom={2}\tDateNaissance={3}".format(
            self.__matricule, self.__nom, self.__prenom, self.__date_naissance
        )

    def __str__(self):
        return self.ToString()


class Ouvrier(Employe):
    SMIG = 2500  # valeur commune à tous les ouvriers

    def __init__(self, matricule, nom, prenom, date_naissance, date_entree):
        super().__init__(matricule, nom, prenom, date_naissance)
        self.__date_entree = date_entree  # string "YYYY-MM-DD" ou date

    def getDateEntree(self):
        return self.__date_entree

    def setDateEntree(self, date_entree):
        self.__date_entree = date_entree

    def __anciennete_en_annees(self):
        """Calcule l'ancienneté en années (approximation simple)"""
        entree = self.__date_to_date(self.__date_entree)
        aujourd_hui = date.today()
        return max(0, aujourd_hui.year - entree.year - ((aujourd_hui.month, aujourd_hui.day) < (entree.month, entree.day)))

    @staticmethod
    def __date_to_date(d):
        """Accepte un date ou un str YYYY-MM-DD"""
        if isinstance(d, date):
            return d
        return datetime.strptime(d, "%Y-%m-%d").date()

    def GetSalaire(self):
        anciennete = self.__anciennete_en_annees()
        salaire = Ouvrier.SMIG + anciennete * 100
        # salaire ne doit pas dépasser SMIG*2
        return min(salaire, Ouvrier.SMIG * 2)

    def ToString(self):
        return "{0}\tDateEntrée={1}\tSalaireMensuel={2}".format(
            super().ToString(), self.__date_entree, self.GetSalaire()
        )

class Cadre(Employe):
    def __init__(self, matricule, nom, prenom, date_naissance, indice):
        super().__init__(matricule, nom, prenom, date_naissance)
        self.__indice = indice  # 1..4

    def getIndice(self):
        return self.__indice

    def setIndice(self, indice):
        self.__indice = indice

    def GetSalaire(self):
        salaires = {
            1: 13000,
            2: 15000,
            3: 17000,
            4: 20000
        }
        return salaires.get(self.__indice, 0)  # 0 si indice invalide

    def ToString(self):
        return "{0}\tIndice={1}\tSalaireMensuel={2}".format(
            super().ToString(), self.__indice, self.GetSalaire()
        )

class Patron(Employe):
    CA = 0 

    def __init__(self, matricule, nom, prenom, date_naissance, pourcentage):
        super().__init__(matricule, nom, prenom, date_naissance)
        self.__pourcentage = pourcentage  

    @classmethod
    def setCA(cls, ca):
        cls.CA = ca

    @classmethod
    def getCA(cls):
        return cls.CA

    def getPourcentage(self):
        return self.__pourcentage

    def setPourcentage(self, pourcentage):
        self.__pourcentage = pourcentage

    def GetSalaire(self):
        return Patron.CA * self.__pourcentage / 100

    def ToString(self):
        return "{0}\tCA={1}\tPourcentage={2}%\tSalaireAnnuel={3}".format(
            super().ToString(), Patron.CA, self.__pourcentage, self.GetSalaire()
        )

if __name__ == "__main__":
    print("=== TEST EXERCICE 4 ===\n")

    # 1) Ouvriers
    ouvrier1 = Ouvrier(1, "Diallo", "Aminata", "1999-01-15", "2018-09-01")
    ouvrier2 = Ouvrier(2, "Traore", "Moussa", "1998-03-10", "2022-01-10")

    # 2) Cadres
    cadre1 = Cadre(3, "Martin", "Nadia", "1990-06-30", 1)
    cadre2 = Cadre(4, "Dupont", "Karim", "1989-02-02", 4)

    # 3) Patrons (CA commun)
    Patron.setCA(1_000_000)  # chiffre d'affaires commun
    patron1 = Patron(5, "Bernard", "Yassine", "1985-11-12", 5)   # 5% du CA
    patron2 = Patron(6, "Aziz", "Salma", "1987-08-21", 12.5)     # 12.5% du CA

    employes = [ouvrier1, ouvrier2, cadre1, cadre2, patron1, patron2]

    for e in employes:
        print(e.ToString())

    print("\n--- Vérifications rapides ---")
    print("Salaire ouvrier1 (capé à SMIG*2):", ouvrier1.GetSalaire())
    print("Salaire cadre2 (indice 4):", cadre2.GetSalaire())
    print("Salaire annuel patron1 (CA*%):", patron1.GetSalaire())
```
