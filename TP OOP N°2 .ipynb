# exercice 1 

```python
# la function factoriel (n)
def factoriel(n):
   
    if n == 0:
        return 1
    resultat = 1
    for i in range(1, n + 1):
        resultat *= i
    return resultat
# question b 
n = int ( input(" enter un nombre j : "))
for j in range (n) : 
    print (factoriel (j) , end= " ")
print() 
```

```python
def icarre (n) : 
    somme  = 0
    impaire  = 1 
    for _ in range (n) :
        somme = somme + impaire
        impaire = impaire + 2 
       
             
    return somme 
#teste 
print (icarre(5))
```

 # exercice 3
 

```python
# function mult0(a,b)
def mult0(a,b) :
    resulta = 0 
    for _ in range (b): 
        resulta = resulta + a
    return resulta 
print ( mult0(2 ,5 ))

#function mult(a,b)
def valeur_absolue ( x) : 
    if x > 0 :
        return x 
    else :
        return -x 
def signe (x) : 
    if x < 0 : 
        return -1 
    elif x == 0 :
        return 0 
    else : 
        return 1 


def mult(a, b):
    signe_resultat = signe(a) * signe(b)
    a_abs = valeur_absolue(a)
    b_abs = valeur_absolue(b)
  
    produit = mult0(a_abs, b_abs)
    if signe_resultat >= 0:
        return produit
    else : 
        return -produit 
        
# (c) Saisie et affichage
a = int(input("Entrez le premier entier : "))
b = int(input("Entrez le second entier : "))

resultat = mult(a, b)
print(f"Le produit de {a} et {b} est : {resultat}")
```

#   exercice 5

```python
def noccurs(lt, e):
    # Vérifier que lt est bien une liste
    if not isinstance(lt, list):
        return -1

    compteur = 0
    for element in lt:
        if element == e:
            compteur += 1

    return compteur
#teste 
print(noccurs(["a", 1, 2, 3, 2, 2, 5, 2], 2))
 
```
