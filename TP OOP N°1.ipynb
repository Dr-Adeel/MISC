```python
from fractions import Fraction 

a , b = 217 , 440 
c , d = 101 , 256 
e , f = 86 , 71 
numerateur = a*d*f + c*b*f + e*b*d
denumerature = b*d*f
pgcd = np.gcd(numerateur , denumerature ) 
new_numerateur = numerateur / pgcd
new_denumerature = denumerature / pgcd 
print (f"alors la forme eruductible est : {new_numerateur / new_denumerature }") 
```

 ## exercice 10

```python

n = int(input("Entrez un entier positif : "))
diviseurs = []

for i in range(1, n + 1):
    if n % i == 0:
        diviseurs.append(i)

diviseurs.sort(reverse=True)

for d in diviseurs:
    print(d, end=" ")

print("==>", len(diviseurs))
```
