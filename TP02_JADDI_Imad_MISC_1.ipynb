# **Q1 : Calcul de factorielle et table des factorielles**

```
def factoriel(n):
    resultat = 1
    for i in range(1, n + 1):
        resultat *= i
    return resultat

n = int(input("Entrez un entier positif : "))

# Affiche seulement les résultats comme dans l'exemple
for i in range(1, n + 1):
    print(factoriel(i), end=" ")
print()

print(f"\nNombre de chiffres de 50! : {len(str(factoriel(50)))}")
```

# **Q2 : Carré par la méthode des impairs**


```
def iarré(n):
    # Calcule le carré de n en additionnant les n premiers nombres impairs
    somme = 0
    for i in range(n):
        somme += 2*i + 1  # Le (i+1)ème nombre impair
    return somme

# Programme principal
if __name__ == "__main__":
    # Test avec les exemples donnés
    tests = [1, 2, 3, 4, 5, 6]

    print("Calcul des carrés par la méthode des impairs:")
    print("=" * 40)

    for n in tests:
        resultat = iarré(n)
        # Afficher le détail du calcul
        details = " + ".join(str(2*i + 1) for i in range(n))
        print(f"{n}² = {details} = {resultat}")


```

# **Q3 : Multiplication sans opérateur de multiplication**

```
# Partie (a): Fonction pour deux entiers positifs
def mult0(a, b):
    if a < 0 or b < 0:
        return "mult0 ne fonctionne qu'avec des entiers positifs"
    resultat = 0
    for _ in range(b):
        resultat += a
    return resultat

# Partie (b): Fonction pour deux entiers quelconques
# D'abord, définissons nos propres fonctions pour signe et valeur absolue
def mon_abs(x):
    if x < 0:
        return -x
    return x

def signe(x):
    if x > 0:
        return 1
    elif x < 0:
        return -1
    else:
        return 0

def mult(a, b):
    # Calculer le produit des valeurs absolues
    abs_a = mon_abs(a)
    abs_b = mon_abs(b)

    # Utiliser mult0 pour calculer le produit des valeurs absolues
    produit_abs = mult0(abs_a, abs_b)

    # Déterminer le signe du résultat
    if signe(a) == signe(b):
        return produit_abs
    else:
        return -produit_abs

# Partie (c): Saisie et affichage
print("Entrez deux entiers :")
a = int(input("Premier entier (a) : "))
b = int(input("Deuxième entier (b) : "))

print("\nRésultats :")
print(f"mult0({a}, {b}) = {mult0(a, b)}")
print(f"mult({a}, {b}) = {mult(a, b)}")
```

# **Q4 : Transformation d'un mot en caractères séparés**

```
def transformer_mot(mot):
    return " ".join(list(mot))

# Tests avec les exemples de l'énoncé
test1 = "altitude"
test2 = "et123"

print(f"'{test1}' → {transformer_mot(test1)}")
print(f"'{test2}' → {transformer_mot(test2)}")

# Version interactive
mot_utilisateur = input("\nEntrez votre propre mot : ")
resultat = transformer_mot(mot_utilisateur)
print(f"Résultat : {resultat}")
```

# **Q5 : Nombre d'occurrences dans une liste**

```
def noccurs(lt, e):
    # Vérifier si lt est bien une liste
    if not isinstance(lt, list):
        return -1

    # Compter les occurrences
    compteur = 0
    for element in lt:
        if element == e:
            compteur += 1
    return compteur

print(f'noccurs(["a",1,2,3,2,2,5,2], 2) = {noccurs(["a",1,2,3,2,2,5,2], 2)}')  # → 4 ✓
print(f"noccurs(2, 2) = {noccurs(2, 2)}")  # → -1 ✓
print(f'noccurs(["a",1,2,3,2,2,5,2], 9) = {noccurs(["a",1,2,3,2,2,5,2], 9)}')  # → 0 ✓
```

# **Q6 : Génération d'une liste de chaînes avec tirets**

---



```
def generer_liste_tirets(n):

    if n <= 0:
        return []

    liste_resultat = []
    for i in range(1, n + 1):
        chaine = "$"
        # Ajouter i tirets
        chaine += "-" * i
        chaine += "$"
        liste_resultat.append(chaine)
    return liste_resultat

# Programme principal pour la question 6
print("\n=== Génération de liste avec tirets ===")
try:
    n = int(input("Entrez un entier n : "))
    resultat = generer_liste_tirets(n)
    print(f"Liste générée : {resultat}")

    # Tests avec les valeurs données
    print("\nTests avec les valeurs données :")
    for test_n in [5, 0, -2, 1]:
        print(f"Pour n = {test_n} : {generer_liste_tirets(test_n)}")

except ValueError:
    print("Veuillez entrer un entier valide !")
```

# **Q7 : Trace d'une matrice**

```
def traceMT(m):
    # Vérifier que m est une matrice carrée
    n = len(m)
    for ligne in m:
        if len(ligne) != n:
            return "Erreur : la matrice n'est pas carrée"

    # Calculer la trace
    trace = 0
    for i in range(n):
        trace += m[i][i]
    return trace

# Programme principal pour la question 7
print("\n=== Calcul de la trace d'une matrice ===")

# Exemple de l'énoncé
matrice_exemple = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(f"Matrice exemple :")
for ligne in matrice_exemple:
    print(ligne)

print(f"Trace de la matrice : {traceMT(matrice_exemple)}")

# Test avec une autre matrice
matrice_test = [
    [2, 0, 0],
    [0, 3, 0],
    [0, 0, 4]
]

print(f"\nMatrice test :")
for ligne in matrice_test:
    print(ligne)

print(f"Trace de la matrice test : {traceMT(matrice_test)}")
```
