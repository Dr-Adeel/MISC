```python
def factoriel(n):
    """Calculates the factorial of a non-negative integer n."""
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Part (b): Saisit un entier positif puis calcule et affiche la table des factoriels
while True:
    try:
        num = int(input("Veuillez saisir un entier positif (ou nul) pour le calcul des factorielles : "))
        if num >= 0:
            break
        else:
            print("L'entier doit être positif ou nul. Veuillez réessayer.")
    except ValueError:
        print("Entrée invalide. Veuillez saisir un nombre entier.")

print(f"\nTable des factorielles jusqu'à {num}! :")
for j in range(num + 1):
    print(f"{j}! = {factoriel(j)}")
```

```python
def icarre(n):
    """Calculates the square of a positive integer n by summing the first n odd numbers."""
    if n <= 0:
        raise ValueError("n must be a positive integer.")

    current_odd = 1
    sum_of_odds = 0
    for _ in range(n):
        sum_of_odds += current_odd
        current_odd += 2  # Move to the next odd number
    return sum_of_odds

# Test cases based on the description
print(f"icarre(1) = {icarre(1)} (Expected: 1)") # 1^2 = 1
print(f"icarre(2) = {icarre(2)} (Expected: 4)") # 2^2 = 1 + 3 = 4
print(f"icarre(3) = {icarre(3)} (Expected: 9)") # 3^2 = 1 + 3 + 5 = 9
print(f"icarre(4) = {icarre(4)} (Expected: 16)") # 4^2 = 1 + 3 + 5 + 7 = 16
print(f"icarre(5) = {icarre(5)} (Expected: 25)") # 5^2 = 1 + 3 + 5 + 7 + 9 = 25
```

```python
def mult0(a, b):
    """Calculates the product of two non-negative integers a and b using repeated addition."""
    if a < 0 or b < 0:
        raise ValueError("mult0 expects non-negative integers.")

    result = 0
    # Add 'a' to 'result', 'b' times
    # Or add 'b' to 'result', 'a' times (potentially more efficient if a < b)
    # To avoid explicit multiplication even for count, we can iterate 'b' times.
    for _ in range(b):
        result += a
    return result

def mult(a, b):
    """
    Calculates the product of two integers a and b without using multiplication
    or predefined abs/sign functions.
    """
    if a == 0 or b == 0:
        return 0

    # Determine the sign of the result
    # The result is negative if one number is positive and the other is negative.
    is_result_negative = False
    if (a > 0 and b < 0) or (a < 0 and b > 0):
        is_result_negative = True

    # Get absolute values without abs()
    abs_a = a if a >= 0 else (0 - a)
    abs_b = b if b >= 0 else (0 - b)

    # Calculate the product of absolute values using mult0
    raw_product = mult0(abs_a, abs_b)

    # Apply the determined sign using subtraction for negation
    if is_result_negative:
        return 0 - raw_product
    else:
        return raw_product

# Test cases based on the description and common scenarios
print(f"Test Case: mult(10, 5) = {mult(10, 5)} (Expected: 50)")
print(f"Test Case: mult(-21, -123) = {mult(-21, -123)} (Expected: 2583)")
print(f"Test Case: mult(10, -5) = {mult(10, -5)} (Expected: -50)")
print(f"Test Case: mult(-10, 5) = {mult(-10, 5)} (Expected: -50)")
print(f"Test Case: mult(0, 7) = {mult(0, 7)} (Expected: 0)")
print(f"Test Case: mult(-8, 0) = {mult(-8, 0)} (Expected: 0)")

```

```python
def noccurs(lt, e):
    """Calculates and returns the number of times an element e appears in a list lt.
    Returns -1 if lt is not a list.
    """
    # Explicitly use the built-in list type to avoid shadowing issues
    if not isinstance(lt, __builtins__.list):
        return -1

    count = 0
    for item in lt:
        if item == e:
            count += 1
    return count

# Test cases
print(f"noccurs([\"a\",1,2,3,2,2,5,2],2) ==> {noccurs(["a",1,2,3,2,2,5,2],2)} (Expected: 4)")
print(f"noccurs(2,2) ==> {noccurs(2,2)} (Expected: -1)")
print(f"noccurs([\"a\",1,2,3,2,2,5,2],9) ==> {noccurs(["a",1,2,3,2,2,5,2],9)} (Expected: 0)")
```

```python
def generate_hyphen_list(n):
    """Génère une liste de chaînes au format ['-','--','---',...] où le dernier élément possède n tirets.
    Renvoie une liste vide si n est négatif ou nul.
    """
    if n <= 0:
        return []

    result_list = []
    for i in range(1, n + 1):
        result_list.append('$' + '-' * i)
    return result_list

# Jeux de tests
print(f"generate_hyphen_list(5) ==> {generate_hyphen_list(5)} (Expected: ['$-$', '$--$', '$---$', '$----$', '$-----$'])")
print(f"generate_hyphen_list(0) ==> {generate_hyphen_list(0)} (Expected: [])")
print(f"generate_hyphen_list(-2) ==> {generate_hyphen_list(-2)} (Expected: [])")
print(f"generate_hyphen_list(1) ==> {generate_hyphen_list(1)} (Expected: ['$-$'])")

```

```python
def traceMT(m):
    """Calculates and returns the trace of a square numeric matrix m.
    The trace is the sum of the elements on the main diagonal.
    """
    # Explicitly use the built-in list type to avoid shadowing issues
    if not isinstance(m, __builtins__.list) or not all(isinstance(row, __builtins__.list) for row in m):
        raise TypeError("Input must be a list of lists representing a matrix.")

    n = len(m)
    if n == 0:
        return 0

    # Check if it's a square matrix and contains only numbers
    for i in range(n):
        if len(m[i]) != n:
            raise ValueError("Matrix must be square (n x n).")
        for j in range(n):
            if not isinstance(m[i][j], (int, float)):
                raise TypeError("Matrix elements must be numeric (int or float).")

    trace_sum = 0
    for i in range(n):
        trace_sum += m[i][i]

    return trace_sum

# Example matrix from the description
matrix_example = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(f"The trace of the matrix {matrix_example} is: {traceMT(matrix_example)} (Expected: 15)")

# Additional test cases
matrix_2x2 = [[10, 20], [30, 40]]
print(f"The trace of the matrix {matrix_2x2} is: {traceMT(matrix_2x2)} (Expected: 50)")

matrix_1x1 = [[7]]
print(f"The trace of the matrix {matrix_1x1} is: {traceMT(matrix_1x1)} (Expected: 7)")

empty_matrix = []
print(f"The trace of the matrix {empty_matrix} is: {traceMT(empty_matrix)} (Expected: 0)")

```
