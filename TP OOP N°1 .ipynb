```python
# Question 1
import numpy as np

def perimetre(rayon):
    return 2 * np.pi * rayon

def aire(rayon):
    return np.pi * rayon ** 2

print("Résultat : 4.56 ==> {:.2f} {:.2f}" .format( aire(4.56),perimetre(4.56)))
```

```python
# Question 4
a = input("Entrer un age: ")
assert a.isdigit(), "Veuillez entrer un entier valide."
match a:
    case a if int(a) <= 5:
        print("OUPS pas suffisamment grand")
    case a if 6 < int(a) <= 9:
        print("Poussin ou Pupille")
    case a if 10 < int(a) <= 16:
        print("Minime ou Cadet")
    case a if 16 < int(a) < 18:
        print("Junior")
    case a if int(a) >= 18:
        print("Senior")
```
