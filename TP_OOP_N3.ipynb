```python
#question 1
class Batiment:
    def __init__(self, adresse=""):
        self.__adresse = adresse

    # ----- Getters -----
    def get_adresse(self):
        return self.__adresse

    # ----- Setters -----
    def set_adresse(self, adresse):
        self.__adresse = adresse

    # ----- ToString -----
    def __str__(self):
        return f"Bâtiment situé à : {self.__adresse}"

```

```python
#question 2
class Maison(Batiment):
    def __init__(self, adresse="", nbPieces=0):
        super().__init__(adresse)
        self.__nbPieces = nbPieces

    # ----- Getters -----
    def get_nbPieces(self):
        return self.__nbPieces

    # ----- Setters -----
    def set_nbPieces(self, nb):
        self.__nbPieces = nb

    # ----- ToString -----
    def __str__(self):
        return f"Maison : {self.get_nbPieces()} pièces, Adresse : {self.get_adresse()}"

```

```python
#question 3
class Test:
    @staticmethod
    def main():
        print("=== Test de la classe Bâtiment ===")
        b1 = Batiment("Rue Hassan II")
        print(b1)

        print("\n=== Test de la classe Maison ===")
        m1 = Maison("Avenue Mohammed V", 5)
        print(m1)

        m2 = Maison()
        m2.set_adresse("Quartier Al Qods")
        m2.set_nbPieces(3)
        print(m2)


# Lancer le test
Test.main()
```

```python
#question 1
#Classe Personne
class Personne:
    def __init__(self, nom, prenom, dateNaissance):
        self.__nom = nom
        self.__prenom = prenom
        self.__dateNaissance = dateNaissance

    # Getters
    def get_nom(self):
        return self.__nom

    def get_prenom(self):
        return self.__prenom

    def get_dateNaissance(self):
        return self.__dateNaissance

    # Méthode polymorphe
    def Afficher(self):
        print(f"Personne : {self.__prenom} {self.__nom}, Né(e) le {self.__dateNaissance}")

#Classe Employé
class Employe(Personne):
    def __init__(self, nom, prenom, dateNaissance, salaire):
        super().__init__(nom, prenom, dateNaissance)
        self.__salaire = salaire

    def get_salaire(self):
        return self.__salaire

    def set_salaire(self, salaire):
        self.__salaire = salaire

    def Afficher(self):
        print(f"Employé : {self.get_prenom()} {self.get_nom()}, Né(e) le {self.get_dateNaissance()}, Salaire : {self.__salaire} DH")


#Classe Chef
class Chef(Employe):
    def __init__(self, nom, prenom, dateNaissance, salaire, service):
        super().__init__(nom, prenom, dateNaissance, salaire)
        self.__service = service

    def get_service(self):
        return self.__service

    def set_service(self, service):
        self.__service = service

    def Afficher(self):
        print(f"Chef : {self.get_prenom()} {self.get_nom()}, Service : {self.__service}, Salaire : {self.get_salaire()} DH")

#Classe Directeur
class Directeur(Chef):
    def __init__(self, nom, prenom, dateNaissance, salaire, service, societe):
        super().__init__(nom, prenom, dateNaissance, salaire, service)
        self.__societe = societe

    def get_societe(self):
        return self.__societe

    def set_societe(self, societe):
        self.__societe = societe

    def Afficher(self):
        print(f"Directeur : {self.get_prenom()} {self.get_nom()}, Société : {self.__societe}, Service : {self.get_service()}, Salaire : {self.get_salaire()} DH")




    
```

```python
#question 1 
from abc import ABC, abstractmethod

class Employe(ABC):
    def __init__(self, matricule, nom, prenom, dateNaissance):
        self._matricule = matricule
        self._nom = nom
        self._prenom = prenom
        self._dateNaissance = dateNaissance

    # ----- Getters -----
    def get_matricule(self):
        return self._matricule

    def get_nom(self):
        return self._nom

    def get_prenom(self):
        return self._prenom

    def get_dateNaissance(self):
        return self._dateNaissance

    # ----- ToString -----
    def __str__(self):
        return f"{self._prenom} {self._nom} (Matricule: {self._matricule})"

    # ----- Méthode abstraite -----
    @abstractmethod
    def GetSalaire(self):
        pass
```

```python
#question 2
#Classe Ouvrier
from datetime import datetime

class Ouvrier(Employe):
    SMIG = 2500

    def __init__(self, matricule, nom, prenom, dateNaissance, dateEntree):
        super().__init__(matricule, nom, prenom, dateNaissance)
        self._dateEntree = dateEntree

    def GetSalaire(self):
        annee_actuelle = datetime.now().year
        anciennete = annee_actuelle - self._dateEntree

        salaire = Ouvrier.SMIG + anciennete * 100

        # Salaire plafonné
        if salaire > Ouvrier.SMIG * 2:
            salaire = Ouvrier.SMIG * 2

        return salaire

    def __str__(self):
        return f"Ouvrier : {super().__str__()}, Entré en {self._dateEntree}, Salaire = {self.GetSalaire()} $"

#Classe Cadre
class Cadre(Employe):
    def __init__(self, matricule, nom, prenom, dateNaissance, indice):
        super().__init__(matricule, nom, prenom, dateNaissance)
        self._indice = indice

    def GetSalaire(self):
        if self._indice == 1:
            return 13000
        elif self._indice == 2:
            return 15000
        elif self._indice == 3:
            return 17000
        elif self._indice == 4:
            return 20000
        else:
            return 0

    def __str__(self):
        return f"Cadre : {super().__str__()}, Indice {self._indice}, Salaire = {self.GetSalaire()} $"

#Classe Patron
class Patron(Employe):
    chiffreAffaire = 0  # commun à tous les patrons

    def __init__(self, matricule, nom, prenom, dateNaissance, pourcentage):
        super().__init__(matricule, nom, prenom, dateNaissance)
        self._pourcentage = pourcentage

    def GetSalaire(self):
        return Patron.chiffreAffaire * self._pourcentage / 100

    def __str__(self):
        return f"Patron : {super().__str__()}, Pourcentage {self._pourcentage}%, Salaire = {self.GetSalaire()} $"
```

```python
#question 3
class Test:
    @staticmethod
    def main():
        # Définir le chiffre d'affaire commun
        Patron.chiffreAffaire = 500000

        personnes = [
            Ouvrier(1, "Ali", "Youssef", "1990-01-01", 2015),
            Ouvrier(2, "Hassan", "Amine", "1985-03-12", 2010),

            Cadre(3, "Karim", "Said", "1992-07-21", 1),
            Cadre(4, "Nadia", "Sara", "1995-11-05", 3),

            Patron(5, "El Idrissi", "Samir", "1970-02-02", 10)
        ]

        for p in personnes:
            print(p)


Test.main()
```
