```python
# Question 1
# a)
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# b)
a = int(input("Enter a number to compute its factorial: "))
print("{} ==> ".format(a),end=" ")
for i in range(1,a+1):
    print("{}".format(factorial(i)),end=" ")

# c) 
print("\n")
print("nombre de chiffres pour {}! : {}".format(a, len(str(factorial(a)))))
```

```python
# Question 2
def icarre(n):
    sum = 0
    for i in range(0, n):
        sum += 2*i + 1
    return sum
print(icarre(9))
```

```python
# Question 3
# a)
def mult0(a,b):
    if b == 0 or a == 0:
        return 0
    else:
        sum = 0
        for i in range(b):
            sum += a
        return sum

# b)
def absval(x):
    if x < 0:
        return -x
    else:
        return x
def mult(a,b):
    if b < 0 or a < 0:
        if a < 0 and b < 0:
            return mult0(absval(a), absval(b))
        else:
            return -mult0(absval(a), absval(b))
    return mult0(a, b)

# c)

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("{} {} ==> {} {} {}".format(a, b, mult(a, b),-mult(a,b),mult(a,b)))
```

```python
# Question 5

def noccurs(liste, numero):
    if not isinstance(liste, list):
        return -1
    n = 0
    for x in liste:
        if x == numero:
            n += 1
    return n
print(noccurs([1, 2, 3, 4, 5, 1, 2, 1, 6, 7, 8, 1], 1))  # Output: 4
```

```python
# Question 7

def traceMT(m):
    return sum(m[i][i] for i in range(len(m)))

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(traceMT(matrix)) 
```
