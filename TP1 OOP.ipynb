##### 2. Écrivez un code qui, sans utiliser le module fractions, trouve la forme irréductible de la somme suivante et qui vérifie le résultat en utilisant ledit module : Pensez à utiliser la fonction gcd (greatest common divisor)

```python

n1, d1 = 217, 440
n2, d2 = 101, 256
n3, d3 = 86, 71

numerateur_total = (n1 * d2 * d3) + (n2 * d1 * d3) + (n3 * d1 * d2)
denominateur_total = d1 * d2 * d3

diviseur = math.gcd(numerateur_total, denominateur_total)
numerateur_simplifie = numerateur_total // diviseur
denominateur_simplifie = denominateur_total // diviseur

print(f"Résultat : {numerateur_simplifie}/{denominateur_simplifie}")

verif = Fraction(217, 440) + Fraction(101, 256) + Fraction(86, 71)
print(f"Vérification : {verif}")
```

##### 10.Écrivez un code qui saisit un entier positif puis qui affiche tous ses diviseurs en ordre décroissant. Comptez également le nombre de diviseurs et affichez-le en tant que dernier nombre de la série.

```python
n = int(input("Entrez un entier positif : "))
diviseurs = []

for i in range(1, int(n/2) + 1):
    if n % i == 0:
        diviseurs.append(i)


diviseurs.sort(reverse=True)
print(n, "==> ", end="")
for d in diviseurs:
    print(d, end=" ")
print(f"=> {len(diviseurs)}")
```
