```python
import tkinter as tk

window = tk.Tk()
window.title("Name Input")
window.minsize(width=300, height=200)

# Label
label = tk.Label(text="Enter your name:")
label.pack()

# Entry (Champ de saisie)
entry = tk.Entry()
entry.pack()

# Fonction pour r√©cup√©rer le texte quand on clique
def get_name():
    name = entry.get()
    print(f"User typed: {name}")

# Bouton de validation
btn = tk.Button(text="Submit", command=get_name)
btn.pack()

window.mainloop()
```

```python
import tkinter as tk

window = tk.Tk()
window.title("Multi-line Text Box")
window.minsize(width=400, height=300)

# Widget Text (pour √©crire plusieurs lignes)
textbox = tk.Text(height=5)
textbox.pack()

# Fonction pour lire le contenu
def read_text():
    # "1.0" signifie : Ligne 1, Caract√®re 0 (le d√©but)
    # tk.END signifie : Jusqu'√† la fin
    content = textbox.get("1.0", tk.END)
    print("--- Contenu de la bo√Æte ---")
    print(content)
    print("---------------------------")

# Bouton pour d√©clencher la lecture
btn = tk.Button(text="Lire le texte", command=read_text)
btn.pack()

# Bouton pour effacer le texte
clear_btn = tk.Button(text="Effacer", command=lambda: textbox.delete("1.0", tk.END))
clear_btn.pack()

window.mainloop()
```

```python
from tkinter import Tk, Label, Button

class MyGUI:
    def __init__(self, master):
        self.master = master
        master.title("A simple GUI")

        self.label = Label(master, text="This is our first GUI within a class!")
        self.label.pack()

        self.greet_button = Button(master, text="Greet", command=self.greet)
        self.greet_button.pack()

        self.close_button = Button(master, text="Close", command=master.quit)
        self.close_button.pack()

    def greet(self):
        print("Greetings from the Class!")

# Lancement propre
root = Tk()
my_gui = MyGUI(root)
root.mainloop()
```

```python
import random
from tkinter import Tk, Label, Button, Entry, StringVar, DISABLED, NORMAL, END, W, E

class GuessingGame:
    def __init__(self, master):
        self.master = master
        master.title("Guess the Number Game")

        self.secret_number = random.randint(1, 100)
        self.attempts = 0
        self.guess = None

        self.message = "Guess a number from 1 to 100"
        self.label_text = StringVar()
        self.label_text.set(self.message)
        
        # Interface
        self.label = Label(master, textvariable=self.label_text)
        
        # Validation (pour emp√™cher d'√©crire des lettres)
        vcmd = master.register(self.validate)
        self.entry = Entry(master, validate="key", validatecommand=(vcmd, '%P'))

        self.guess_button = Button(master, text="Guess", command=self.guess_number)
        self.reset_button = Button(master, text="Play Again!", command=self.reset, state=DISABLED)

        # Layout (Grid)
        self.label.grid(row=0, column=0, columnspan=2, sticky=W+E)
        self.entry.grid(row=1, column=0, columnspan=2, sticky=W+E)
        self.guess_button.grid(row=2, column=0)
        self.reset_button.grid(row=2, column=1)

    def validate(self, new_text):
        if not new_text:
            self.guess = None
            return True
        try:
            self.guess = int(new_text)
            # On v√©rifie si c'est entre 1 et 100
            if self.guess < 1 or self.guess > 100:
                return False 
            return True
        except ValueError:
            return False
        
    def guess_number(self):
        self.attempts += 1
        if self.guess is None:
            self.message = "Please enter a number first."
        elif self.guess == self.secret_number:
            suffix = '' if self.attempts == 1 else 's'
            self.message = f"Win! The number was {self.secret_number} found in {self.attempts} attempt{suffix}!"
            self.guess_button.configure(state=DISABLED)
            self.reset_button.configure(state=NORMAL)
        elif self.guess < self.secret_number:
            self.message = "Too low! Guess again!"
        else:
            self.message = "Too high! Guess again!"
        
        self.label_text.set(self.message)

    def reset(self):
        self.entry.delete(0, END)
        self.secret_number = random.randint(1, 100)
        self.attempts = 0
        self.guess = None
        self.message = "Guess a number from 1 to 100"
        self.label_text.set(self.message)
        self.guess_button.configure(state=NORMAL)
        self.reset_button.configure(state=DISABLED)

# Lancement du jeu
root = Tk()
game = GuessingGame(root)
root.mainloop()
```

```python
import tkinter as tk

class ComputerGuessGame:
    def __init__(self, master):
        self.master = master
        master.title("L'Ordinateur devine votre nombre")
        master.geometry("400x200")

        # --- Variables du jeu ---
        self.min_val = 1
        self.max_val = 100
        self.current_guess = 50  # On commence au milieu
        
        # --- Interface Graphique ---
        
        # Instruction
        self.instruction_label = tk.Label(master, text="Pensez √† un nombre entre 1 et 100.", font=("Arial", 12))
        self.instruction_label.pack(pady=10)

        # La question de l'ordinateur
        self.question_label = tk.Label(master, text=f"Est-ce que c'est {self.current_guess} ?", font=("Arial", 16, "bold"), fg="blue")
        self.question_label.pack(pady=10)

        # Frame pour aligner les boutons
        self.button_frame = tk.Frame(master)
        self.button_frame.pack(pady=10)

        # Les 3 boutons de r√©ponse
        # "C'est plus petit" (donc l'ordi a vis√© trop haut)
        self.btn_lower = tk.Button(self.button_frame, text="C'est plus petit", command=self.go_lower, bg="#ffcccb")
        self.btn_lower.pack(side=tk.LEFT, padx=10)

        # "C'est √ßa !" (Victoire)
        self.btn_correct = tk.Button(self.button_frame, text="C'est √ßa !", command=self.success, bg="#ccffcc")
        self.btn_correct.pack(side=tk.LEFT, padx=10)

        # "C'est plus grand" (donc l'ordi a vis√© trop bas)
        self.btn_higher = tk.Button(self.button_frame, text="C'est plus grand", command=self.go_higher, bg="#ccf2ff")
        self.btn_higher.pack(side=tk.LEFT, padx=10)

    def next_guess(self):
        """Calcule le prochain nombre"""
        if self.min_val > self.max_val:
            self.question_label.config(text="Vous trichez ! ü§î", fg="red")
            self.disable_buttons()
            return

        # Calcul savant (division enti√®re par 2)
        self.current_guess = (self.min_val + self.max_val) // 2
        self.question_label.config(text=f"Est-ce que c'est {self.current_guess} ?")

    def go_lower(self):
        """L'utilisateur dit que son nombre est plus PETIT que la proposition"""
        # Donc le maximum possible devient le nombre propos√© - 1
        self.max_val = self.current_guess - 1
        self.next_guess()

    def go_higher(self):
        """L'utilisateur dit que son nombre est plus GRAND que la proposition"""
        # Donc le minimum possible devient le nombre propos√© + 1
        self.min_val = self.current_guess + 1
        self.next_guess()

    def success(self):
        """L'ordinateur a trouv√©"""
        self.question_label.config(text=f"J'ai gagn√© ! C'√©tait {self.current_guess} !", fg="green")
        self.instruction_label.config(text="Merci d'avoir jou√©.")
        self.disable_buttons()

    def disable_buttons(self):
        self.btn_lower.config(state=tk.DISABLED)
        self.btn_higher.config(state=tk.DISABLED)
        self.btn_correct.config(state=tk.DISABLED)

# --- Lancement du programme ---
root = tk.Tk()
app = ComputerGuessGame(root)
root.mainloop()
```
