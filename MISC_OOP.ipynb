## Declaration and Definition of Classes and their Instances
- Attributs (Properties (Attribut values are the data), State)
- Methods (Operations, Behavior)



```python
#Object Oriented Programming
# Encapsulation

class Vehicle:
  specificity = "Testing Vehicles"
  instances = []

  def __init__(self, model: str, year:int, price: float, availability:bool):

    assert price >= 0, f"{price} is not a valid price."
    assert year <= 2025, f"{year} is not a valid year."

    self.model = model
    self.price = price
    self.year = year
    self. availability = availability

    Vehicle.instances.append(self)

  def __str__(self):
    return f"{self.model} is a Vehicle"

  def detail(self, color="Green"):
    return f"{self.model} has color {color}"

  def describe(self):
    return f"{self.model} is {self.year} of production."


```

```python
# Instances (Delaration of Objects)
Vehicule = Vehicle("Tesla", 2024, 38000, False)

Vehicule.describe()
```

##Inheritance and Polymorphism/Overloading

```python
class Truck(Vehicle):
  def capacity(self, cap:int = 6000):
    return f"{self.model} has capacity {cap}"

  def detail(self, color="Black"):
    return super().detail(color)

class Car(Vehicle):
  def detail(self, color="White"):
    return super().detail(color)
```

```python
## Instances of child classes


Van = Truck("Toyota", 2008, 15000, True)
Mercedese = Car("Mercedese", 2015, 30000, True)
```
