# **polymorphism ğŸ­**

```
from abc import ABC, abstractmethod

class Shape:
    @abstractmethod
    def area(self):
        pass
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14*self.radius**2
class Square(Shape):
    def __init__(self,side):
        self.side = side
    def area(self):
        return self.side**2

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height
    def area(self):
        return self.base*self.height*0.5
shapes = [Circle(4), Square(5), Triangle(6, 7)]
for shape in shapes:
    print(f"{shape.area()} cm^2")
```

```
class Library:

  def __init__(self, name):
    self.name = name
    self.books = []

  def add_book(self, book):
    self.books.append(book)

  def list_books(self):
    return [f"{book.title}  by {book.author}" for book in self.books]

class Book:
  def __init__(self, title, author):
    self.title = title
    self.author = author
library = Library("New york")
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald")
book2 = Book("To Kill a Mockingbird", "Harper Lee")
book3 = Book("1984", "George Orwell")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

print(library.name)

for book in library.list_books():
  print(book)
```

# **magic methods ğŸŒŸ**

```
class Book:
    def __init__(self, title, author, num_pages):
      self.title = title
      self.author = author
      self.num_pages = num_pages

    def __str__(self):
      return f"'{self.title}' by {self.author}"

    def __eq__(self,other):
      return self.title == other.title and self.author == other.author

    def __lt__(self,other):
      return self.num_pages < other.num_pages

    def __gt__(self,other):
      return self.num_pages > other.num_pages

    def __add__(self,other):
      return f"{self.num_pages + other.num_pages} pages"

    def __le__(self,other):
      return self.num_pages <= other.num_pages

    def __contains__(self, word):
        return (
            word.lower() in self.title.lower()
            or word.lower() in self.author.lower()
            or word == str(self.num_pages)
        )
    def __getitem__(self, key):
        if key == "title":
            return self.title
        elif key == "author":
            return self.author
        elif key == "num_pages":
            return self.num_pages
        else:
            return f"Key '{key}' was not found"


book1 = Book("The Alchemist", "kayle netro", 192)
book2 = Book("Harry Potter and The Philosopher", "J.K.Rowling", 565)
book3 = Book("The Alchemist", "Paulo Coelho", 788)

print(book1)
print(book2)
print(book3)
print("-"*40)
print(book2 == book1)
print(book1 < book3)
print(book1 > book3)
print(book1 + book3)
print(book1 <= book3)
print('788' in book3)
print(book3['author'])
print(book2['num_pages'])
print(book2['coscos'])


```

# **@property decorator âš™ï¸**

```
class Rectangle:

    def __init__(self, width, height):
      self._width = width
      self._height = height

    @property
    def width(self):
      return f"{self._width :.1f} cm"

    @property
    def height(self):
      return f"{self._height :.1f} cm"

    @width.setter
    def width(self, new_width):
      if new_width > 0:
        self._width = new_width
      else:
        print("Width must be greater than zero")

    @height.setter
    def height(self, new_height):
        if new_height > 0:
            self._height = new_height
        else:
            print("Height must be greater than zero")

    @width.deleter
    def width(self):
      del self._width
      print("Width has been deleted")

    @height.deleter
    def height(self):
      del self._height
      print("Height has been deleted")

rectangle = Rectangle(3, 4)

rectangle.width = 5
rectangle.height = 6

del rectangle.width
del rectangle.height

#print(rectangle.width)
#print(rectangle.height)
```

```
# === CLASSES DE BASE ===
class TAdresse:
    def __init__(self, rue, ville, code_postal):
        self.rue = rue
        self.ville = ville
        self.code_postal = code_postal

class TEtude:
    def __init__(self, domaine, niveau):
        self.domaine = domaine
        self.niveau = niveau

class TRib:
    def __init__(self, iban, bic):
        self.iban = iban
        self.bic = bic

# === PERSONNE ===
class Personne:
    def __init__(self, nom, prenoms, adresse):
        self.nom = nom
        self.prenoms = prenoms
        self.adresse = adresse

    def afficher(self):
        print(f"Nom: {self.nom}")
        print(f"PrÃ©noms: {' '.join(self.prenoms)}")

    def nouvelle_adresse(self, nvadr):
        self.adresse = nvadr

# === Ã‰TUDIANT ===
class Etudiant(Personne):
    def __init__(self, nom, prenoms, adresse, numeroEtudiant, dateNaissance, etudes):
        super().__init__(nom, prenoms, adresse)
        self.numeroEtudiant = numeroEtudiant
        self.dateNaissance = dateNaissance
        self.etudes = etudes
        self.validations = []  # Liste des validations

    def afficher(self):
        super().afficher()
        print(f"NumÃ©ro Ã©tudiant: {self.numeroEtudiant}")

# === ENSEIGNANT ===
class Enseignant(Personne):
    def __init__(self, nom, prenoms, adresse, statut, rib):
        super().__init__(nom, prenoms, adresse)
        self.statut = statut
        self.rib = rib

    def Operation(self):
        print("OpÃ©ration effectuÃ©e")

# === DOCUMENT ===
class Document:
    def __init__(self, titre):
        self.titre = titre

# === VALIDATION ===
class Validation:
    def __init__(self, note, annee):
        self.note = note
        self.annee = annee
        self.etudiant = None
        self.document = None

    def ajouter_etudiant(self, etudiant):
        self.etudiant = etudiant
        etudiant.validations.append(self)

    def ajouter_document(self, document):
        self.document = document

# === COURS ===
class Cours:
    def __init__(self, nom, cycle):
        self.nom = nom
        self.cycle = cycle
        self.objectifs = []
        self.questions = []
        self.pre_requis = []

    def ajouter_objectif(self, objectif):
        self.objectifs.append(objectif)

    def ajouter_question(self, question):
        self.questions.append(question)

    def ajouter_pre_requis(self, cours):
        self.pre_requis.append(cours)

# === OBJECTIF ===
class Objectif:
    def __init__(self, contenu):
        self.contenu = contenu

# === QUESTION ===
class Question:
    def __init__(self, enonce):
        self.enonce = enonce

# === UTILISATION ===
if __name__ == "__main__":
    # CrÃ©ation
    adresse = TAdresse("123 Rue Exemple", "Paris", "75000")
    etudes = TEtude("Informatique", "Master")

    # Ã‰tudiant
    etudiant = Etudiant("Dupont", ["Jean"], adresse, 12345, "2000-05-15", etudes)

    # Document
    document = Document("MÃ©moire de recherche")

    # Validation
    validation = Validation(16, "2024")
    validation.ajouter_etudiant(etudiant)
    validation.ajouter_document(document)

    # Cours
    cours = Cours("Programmation POO", 2)
    objectif = Objectif("Comprendre les classes")
    question = Question("Qu'est-ce que l'encapsulation?")

    cours.ajouter_objectif(objectif)
    cours.ajouter_question(question)

    # Affichage
    etudiant.afficher()
    print(f"Document: {document.titre}")
    print(f"Note: {validation.note}")
    print(f"Cours: {cours.nom}")
```
