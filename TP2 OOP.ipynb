##### Le factoriel d'un entier positif ùëõ est : ùëõ!=Œ†ùëõùëó=1ùëó avec 0!=1.
##### √âcrivez un code qui :


##### (a) D√©finit une fonction factoriel(n) qui calcule et renvoie ùëõ! (avec ùëõ suppos√© positif ou nul).


```python
def factoriel(n):
    r=1
    if n == 0 or n == 1:
        return r
    for i in range(2, n + 1):
        r *= i
    return r
```

##### (b) Saisit un entier positif puis calcule et affiche la table des factoriels des nombres j entre 0 et l'entier saisi.


```python
n = int(input("Entrez un entier positif n : "))
print(n,"==> ",end= " ")
for j in range(1,n + 1):
    print(factoriel(j),end=" ")
```

##### (c) Combien de chiffres faut-il pour √©crire l'entier 50! ? (R√©sultat: 65 chiffres)

```python
nombre_chiffres = len(str(factoriel(50)))
print(" pour 50! : ",nombre_chiffres," chiffres")
```

##### 2. √âcrivez un code qui d√©finit une fonction icarre(n) qui calcule et renvoie le carr√© d'un entier positif n par la m√©thode des impairs, laquelle consiste √† faire la somme des ùëõn premiers nombres impairs.
###### ‚Ä¢ 12=1 : addition du premier impair
###### ‚Ä¢ 22=1+3=4 : addition des deux premiers impairs
###### ‚Ä¢ 32=1+3+5=9 : addition des trois premiers impairs
###### ‚Ä¢ 42=1+3+5+7=16 : addition des quatre premiers impairs
###### ‚Ä¢ 52=1+3+5+7+9=25 : addition des cinq premiers impairs
###### ‚Ä¢ Et ainsi de suite

```python
def icarre(n):
    somme = 0
    for i in range(n):
        somme += 2 * i + 1  
    return somme

    
n = int(input("\nEntrez un entier positif : "))
print(f"{n}¬≤ = {icarre(n)} (verification : {n**2})")
```

##### 3. √âcrivez un code qui :
##### (a) D√©finit une fonction mult0(a,b) qui calcule et renvoie le produit de deux entiers positifs a et b sans faire appel √† l'op√©rateur de multiplication.

```python
def mult0(a, b):
    resultat = 0
    for _ in range(b):
        resultat += a
    return resultat
```

##### (b) D√©duisez une fonction mult(a,b) qui calcule et renvoie le produit de deux entiers a et b sans faire appel aux fonctions pr√©d√©finies (signe ou valeur-absolue -- vous devez les red√©finir).

```python
def valeur_absolue(x):
    return -x if x < 0 else x
def signe(x):
    if x > 0:
        return 1
    elif x < 0:
        return -1
    else:
        return 0

def mult(a, b):

    abs_a = valeur_absolue(a)
    abs_b = valeur_absolue(b)
    resultat_abs = mult0(abs_a, abs_b)
    
    sgn = signe(a) * signe(b)
    
    if sgn >= 0:
        return resultat_abs
    else:
        return -resultat_abs
```

##### (c) Saisissez deux entiers puis affichez le r√©sultat.

##### 5. √âcrivez un code qui d√©finit une fonction noccurs(lt,e) qui calcule et renvoie le nombre de fois qu'un √©l√©ment e appara√Æt dans une liste lt, et ‚àí1 si lt n'a pas le bon type.


```python
def noccurs(lt, e):
    if not isinstance(lt, list):
        return -1
    
    compte = 0
    for element in lt:
        if element == e:
            compte += 1
    return compte

print(noccurs(["a",1,2,3,2,2,5,2],2))
print(noccurs(2,2))
print(noccurs(["a",1,2,3,2,2,5,2],9))
```

##### √âcrivez un code qui d√©finit une fonction traceMT(m) qui calcule et renvoie la trace d'une matrice num√©rique (int ou float) m de taille n√ón des valeurs trace(M)=Œ£j=1..nMjj c.-√†-d. la somme des valeurs de la premi√®re diagonale de la matrice. Exemple : La trace de la matrice ci-dessous est 15 (somme des valeurs 1, 5 et 9 qui forment la premi√®re diagonale) :
#### 1 2 3
#### 4 5 6
#### 7 8 9

```python
def traceMT(m):
    # Si matrice pas carr√©e 
    n = len(m)
    for ligne in m:
        if len(ligne) != n:
            raise ValueError("La matrice n'est pas carr√©e")
    
    trace = 0
    for i in range(n):
        trace += m[i][i]  
    return trace

matrice_ex = [
    [5, 2, 3],
    [4, 3, 6],
    [7, 8, 0]
]

print(matrice_ex)
print(f"\nTrace de la matrice : {traceMT(matrice_ex)}")

```
