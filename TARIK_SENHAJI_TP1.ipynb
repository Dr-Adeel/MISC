### Exercice 2

```python
import math
from fractions import Fraction 


n1, d1 = 217, 440
n2, d2 = 101, 256
n3, d3 = 86, 71

num_inter = (n1 * d2) + (n2 * d1)
den_inter = d1 * d2

num_final = (num_inter * d3) + (n3 * den_inter)
den_final = den_inter * d3

diviseur_commun = math.gcd(num_final, den_final)
num_reduit = num_final // diviseur_commun
den_reduit = den_final // diviseur_commun

print(f"Résultat : {num_reduit}/{den_reduit}")

verif = Fraction(217, 440) + Fraction(101, 256) + Fraction(86, 71)
print("Verification {}".format(verif))
```

### Exercice 4

```python
age = int(input("entrez l'age de l'enfant : "))

if 0 <= age <= 5:
    print("pas suffisamment grand") 
elif 6 <= age <= 9:
    print("poussin ou pupille") 
elif 10 <= age <= 15:
    print("minime ou cadet") 
elif 16 <= age < 18:
    print("junior") 
elif age >= 18:
    print("senior") 
```

### Exercice 6

```python
jour = int(input("Jour (0=Dim...6=Sam) : "))
heure = int(input("Heure (0-24) : "))

est_semaine = (jour >= 1 and jour <= 5)
est_samedi = (jour == 6)
matin = (heure >= 8 and heure < 13)
apres_midi = (heure >= 14 and heure < 17)

ouvert = False

if est_semaine:
    if matin or apres_midi:
        ouvert = True
elif est_samedi:
    if matin: 
        ouvert = True

if ouvert:
    print("Vrai")
else:
    print("Faux")
```

### Exercice 10

```python
nombre = int(input("saisir votre nombre postive : "))
liste = []
compteur = 0 
for i in range(nombre , 0 , -1):
    if nombre % i == 0:
        liste.append(i)
        compteur = compteur + 1

print("nombre {} ==> {} ==> {}".format(nombre , liste , compteur))
```
