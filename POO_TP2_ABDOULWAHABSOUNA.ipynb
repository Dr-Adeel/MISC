```python
# Exercice 1: Calcul de factoreil (basique)
```

```python
def factoriel(n):
    fact = 1
    if n == 0 or n==1 :
        return 1
    else:
        for i in range(1,n+1):
            fact = fact*i
            i= i+1
    return fact
```

```python
n = int(input("Entrez un entier positif : "))
for j in range(0, n + 1):
    print(factoriel(j), end=" ")
print()
f50 = factoriel(50)
print("Nombre de chiffres de 50! :", len(str(f50)))
```

La fonction parcourt les impairs et les additionne pour obtenir n carr√©

```python
def icarre(n):
    somme = 0
    impair = 1
    for _ in range(n):
        somme += impair
        impair += 2
    return somme
n = int(input("Entrez un entier positif pour le carr√© : "))
print("Carr√© par la m√©thode des impairs :", icarre(n))
print("V√©rification :", n * n)
```

```python
def mult0(a, b):
    resultat = 0
    for _ in range(b):
        resultat += a
    return resultat

def valeur_absolue(x):
    if x < 0:
        return -x
    return x

def signe(x):
    if x < 0:
        return -1
    elif x > 0:
        return 1
    return 0
def mult(a, b):
    sa = signe(a)
    sb = signe(b)
    aa = valeur_absolue(a)
    bb = valeur_absolue(b)
    return sa * sb * mult0(aa, bb)
```

```python
def noccurs(lt, e):
    if type(lt) != list:
        return -1
    compteur = 0
    for x in lt:
        if x == e:
            compteur += 1
    return compteur
print(noccurs(["a", 1, 2, 3, 2, 2, 5, 2], 2))   # 4
print(noccurs(2, 2))                         # -1
print(noccurs(["a", 1, 2, 3, 2, 2, 5, 2], 9))
```

```python
def genere_liste(n):
    if n <= 0:
        return []
    liste = []
    for i in range(1, n + 1):
        chaine = "$" + "-" * i + "$"
        liste.append(chaine)
    return liste
print(genere_liste(5))   
print(genere_liste(0))   
print(genere_liste(-2))  
print(genere_liste(1))  
```

```python
def traceMT(m):
    n = len(m)
    somme = 0
    for i in range(n):
        somme += m[i][i]
    return somme
```

```python
M = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Trace de la matrice :", traceMT(M)) 
```
