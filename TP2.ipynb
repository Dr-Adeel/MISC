### Exercice 1
Calcul de la factorielle d'un nombre entier positif et affichage du nombre de chiffres nécessaires pour écrire 50!.

```python
def factoriel(n):
    res = 1
    for i in range(1, n + 1):
        res = res * i
    return res

valeur = int(input("Entrez un entier positif : "))

print("Table des factoriels :")
for j in range(valeur + 1):
    print(factoriel(j), end=" ")
print() 

grand_nombre = factoriel(50)
taille = len(str(grand_nombre))
print("Il faut", taille, "chiffres pour écrire 50")
```

### Exercice 2
Calcul du carré d'un entier `n` en utilisant la propriété que le carré est la somme des `n` premiers nombres impairs.

```python
def icarre(n):
    somme = 0
    impair = 1
    
    for i in range(n):
        somme = somme + impair
        impair = impair + 2 
        
    return somme

print("Le carré de 5 est :", icarre(5)) 
```

### Exercice 3
Implémentation de la multiplication de deux entiers (positifs ou négatifs) en utilisant uniquement l'addition.

```python
def mult0(a, b):
    resultat = 0
    for i in range(b):
        resultat = resultat + a
    return resultat

def mult(a, b):
    signe_final = 1
    
    if a < 0:
        signe_final = -signe_final
        a = -a 
        
    if b < 0:
        signe_final = -signe_final
        b = -b

    prod = mult0(a, b)
    
    if signe_final == -1:
        return -prod
    else:
        return prod

v1 = int(input("Entrez le premier entier : "))
v2 = int(input("Entrez le deuxième entier : "))
print("Résultat :", mult(v1, v2))
```

### Exercice 4
Programme qui prend un mot en entrée et affiche ses lettres séparées par des espaces.

```python
mot = input("Entrez un mot : ")

liste_caracteres = list(mot)

resultat = ""
for lettre in liste_caracteres:
    resultat = resultat + lettre + " "

print(resultat)
```

### Exercice 5
Fonction `noccurs` qui compte le nombre d'occurrences d'un élément `e` dans une liste `lt`.

```python
def noccurs(lt, e):
    if type(lt) != list:
        return -1
    
    compteur = 0
    for element in lt:
        if element == e:
            compteur = compteur + 1
            
    return compteur

print(noccurs(["a", 1, 2, 3, 2, 2], 2)) 
print(noccurs(123, 2))                 
```

### Exercice 6
Génération d'une liste contenant des chaînes de caractères de longueur croissante composées de tirets.

```python
n = int(input("Entrez un entier n : "))
liste_finale = []

for i in range(1, n + 1):
    chaine = "-" * i
    liste_finale.append(chaine)

print(liste_finale)
```

### Exercice 7
Calcul de la trace d'une matrice carrée (somme des éléments de la diagonale principale).

```python
def traceMT(m):
    somme = 0
    taille = len(m) 
    
    for i in range(taille):
        somme = somme + m[i][i]
        
    return somme

matrice = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Trace de la matrice :", traceMT(matrice)) 
```
