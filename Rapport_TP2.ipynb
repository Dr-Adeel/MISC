#### Exercice 1 : Calcul de factoriels

##### L'objectif :

Apprendre à créer une fonction pour calculer le factoriel d'un nombre (n!) et manipuler des grands nombres en Python.

##### Ma stratégie de résolution :

* La fonction : J'ai créé factoriel(n) qui utilise une boucle pour multiplier les nombres de 1 à n.
* La table : Le programme demande un entier et affiche le factoriel de chaque nombre jusqu'à cette limite.
* Grands nombres : Pour 50!, on transforme le résultat en texte (string) pour compter facilement le nombre de chiffres avec len().

```python
# Question a : Définition de la fonction
def factoriel(n):
    if n == 0 or n == 1:
        return 1
    f = 1
    for i in range(1, n + 1):
        f = f * i
    return f

# Question b : Affichage de la table
nb = int(input("Saisir un entier : "))
for i in range(1, nb + 1):
    print(factoriel(i), end=" ")

# Question c : Nombre de chiffres pour 50!
f50 = factoriel(50)
print(f"\nNombre de chiffres de 50! : {len(str(f50))}")
```

##### Points clés de l'exercice :

* Boucle de multiplication : Utiliser une variable qui accumule le produit est la base du calcul de factoriel.
* Conversion de type : Transformer un nombre en chaîne de caractères est une astuce rapide pour mesurer sa longueur.


#### Exercice 2 : Carré par la méthode des impairs

##### L'objectif :

Calculer le carré d'un nombre sans utiliser l'opérateur de puissance, mais en additionnant les nombres impairs.

##### Ma stratégie de résolution :

* Logique mathématique : Pour un nombre n, on doit additionner les n premiers nombres impairs (1, 3, 5, etc.).
* La boucle : J'utilise range(1, 2*n, 2) pour obtenir exactement les nombres impairs nécessaires et les ajouter à la somme.

```python
def icarree(n):
    s = 0
    for i in range(1, 2 * n, 2):
        s = s + i
    return s

print(f"Le carré de 5 est : {icarree(5)}")
```

##### Points clés de l'exercice :

* Range avec un pas : Le pas de 2 dans la fonction range est très utile pour sauter les nombres pairs.
* Algorithme alternatif : On voit qu'il existe souvent plusieurs méthodes mathématiques pour arriver au même résultat informatique.


#### Exercice 3 : Multiplication sans opérateur '*'

##### L'objectif :

Multiplier deux nombres (positifs ou négatifs) en utilisant uniquement des additions.

##### Ma stratégie de résolution :

* Valeur absolue : J'ai créé une fonction inv(n) pour gérer les nombres sans s'occuper du signe au début.
* Addition répétée : On ajoute a à lui-même b fois pour simuler la multiplication.
* Gestion du signe : À la fin, si un seul des deux nombres est négatif, le résultat devient négatif.

```python
def inv(n):
    return -n if n < 0 else n

def mult(a, b):
    p = 0
    for i in range(0, inv(b)):
        p = p + inv(a)
    if (a < 0 and b > 0) or (a > 0 and b < 0):
        return -p
    return p

n1 = int(input("Premier nombre : "))
n2 = int(input("Second nombre : "))
print(f"Résultat {n1}x{n2} = {mult(n1, n2)}")
```

##### Points clés de l'exercice :

* Règle des signes : Il est important de séparer le calcul de la valeur et la détermination du signe final.
* Logique de base : Cet exercice montre comment les opérations complexes sont construites à partir d'opérations simples.


#### Exercice 4 : Transformation de mot en liste

##### L'objectif :

Saisir un mot, le diviser en caractères et l'afficher avec des espaces.

##### Ma stratégie de résolution :
* Conversion : La fonction list(mot) transforme directement une chaîne en une liste de lettres.
* Reconstruction : J'utilise une boucle pour parcourir cette liste et ajouter chaque lettre à une nouvelle chaîne avec un espace.

```python
mot = input("Entrer un mot : ")
liste_char = list(mot)

chf = ""
for i in liste_char:
    chf = chf + i + " "
print(chf)
```

#### Exercice 5 : Comptage d'occurrences dans une liste

##### L'objectif :

Créer une fonction qui compte combien de fois un élément se trouve dans une liste et vérifie le type de l'entrée.

##### Ma stratégie de résolution :

* Vérification du type : On utilise type(lt) != list pour renvoyer -1 si l'utilisateur ne donne pas une liste.
* Compteur : On parcourt la liste et on ajoute 1 à chaque fois que l'élément est trouvé.

```python
def noccurs(lt, e):
    if type(lt) != list:
        return -1
    c = 0
    for i in lt:
        if i == e:
            c += 1
    return c

print(noccurs(["a", 1, 2, 3, 2, 2, 5, 2], 2))
```

##### Points clés de l'exercice :

* Sécurité du code : Vérifier le type des données avant de travailler évite des erreurs de programme.
* Recherche simple : C'est l'algorithme de base pour filtrer ou compter des informations.


#### Exercice 6 : Génération de motifs textuels

##### L'objectif :

Générer une liste de chaînes de caractères avec un motif qui grandit (nombre de tirets).

##### Ma stratégie de résolution :

* Boucle de création : On utilise une boucle de 1 à n pour créer chaque élément.
* Répétition de caractères : En Python, i * "-" permet de répéter le tiret plusieurs fois très facilement.

```python
n = int(input("Entrer un nombre entier : "))
lista = []
for i in range(1, n + 1):
    motif = "$" + i * "-" + "$"
    lista.append(motif)
print(lista)
```

##### Points clés de l'exercice :

* Multiplication de strings : Une fonctionnalité de Python très pratique pour créer des graphiques ou des motifs simples.
* Gestion des listes : Utiliser append() pour construire une liste dynamiquement.


#### Exercice 7 : Trace d'une matrice

##### L'objectif :

Calculer la somme des éléments de la diagonale principale d'une matrice carrée.
+1

##### Ma stratégie de résolution :

* Double boucle : On parcourt les lignes (i) et les colonnes (j) de la matrice.
* Condition de diagonale : La diagonale correspond aux éléments où le numéro de la ligne est égal au numéro de la colonne (i==j).

```python
def traceMT(m):
    s = 0
    for i in range(len(m)):
        for j in range(len(m[i])):
            if i == j:
                s = s + m[i][j]
    return s

matrice = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(f"La trace est : {traceMT(matrice)}")
```

##### Points clés de l'exercice :

* Matrices en Python : Une matrice est représentée par une liste de listes.
* Repérage spatial : Comprendre que M[i][j] avec i=j définit la diagonale est essentiel pour l'algèbre linéaire en informatique.
