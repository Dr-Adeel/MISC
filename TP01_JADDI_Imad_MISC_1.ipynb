# **Q2 : Somme de fractions**

**Description**

Calcul de la somme de plusieurs fractions données sous forme de listes de tuples (numérateur, dénominateur).

**Code utilisé**

```
def somme_fractions(fractions):

    nom = 0  # numérateur total
    den = 1  # dénominateur total

    for fraction in fractions:
        nom += fraction[0]
        den *= fraction[1]
    return nom, den
def afficher_resultat(fractions):

    somme = somme_fractions(fractions)
    print(f"La somme des fractions est : {somme[0]}/{somme[1]}")
# Exemple d'utilisation
fractions = [(217, 440), (101, 256), (86, 71)]
afficher_resultat(fractions)
```

# **Q3 : Maximum de trois nombres**

**Description**

Détermination du plus grand de trois nombres à l’aide d’une classe.

```
def trouver_maximum(a, b, c):

    return max(a, b, c)

# Liste de tests
tests = [
    (8, 6, 8),
    (10, 6, 1),
    (10, 4, -5),
    (5, 0, 1),
    (3, -9, -4)
]

# Résultats des tests
for a, b, c in tests:
    print(f"{a}, {b}, {c} Nombre grand est {trouver_maximum(a, b, c)}")
```

# **Q4 : Catégorie sportive selon l’âge**

**Description**

Détermination de la catégorie sportive d’un individu en fonction de son âge.

**Code utilisé**

```
def determiner_categorie(age):

    if age <= 5:
        return "OUPS pas suffisamment grand"
    elif 6 <= age <= 9:
        return "Poussin ou Pupille"
    elif 10 <= age <= 15:
        return "Minime ou Cadet"
    elif 16 <= age <= 17:
        return "Junior"
    else:
        return "Senior"

# Liste d'âges pour tester
ages_tests = [7, 10, 12, 16, 18, 20, 5]

# Affichage des résultats
for age in ages_tests:
    print(f"{age} → {determiner_categorie(age)}")
```

# **Q6 : Horaire d’ouverture d’un guichet**

**Description**

Vérification de l’ouverture d’un guichet en fonction du jour et de l’heure.

**Code utilisé**

```
def est_ouvert(jour, heure):

    # Dimanche toujours fermé
    if jour == 0:
        return False

    # Samedi uniquement le matin (8h-13h)
    if jour == 6:
        return 8 <= heure < 13

    # Lundi à vendredi : matin (8h-13h) et après-midi (14h-17h)
    matin = 8 <= heure < 13
    apres_midi = 14 <= heure < 17

    return matin or apres_midi

# Tests
tests = [
    (1, 9),
    (0, 5),
    (6, 15),
    (6, 13),
    (3, 16)
]

# Affichage des résultats
for jour, heure in tests:
    print(f"{jour} {heure} ==> {est_ouvert(jour, heure)}")
```

# **Q7 : Détermination de la saison**

**Description**

Détermination de la saison (hiver, printemps, été, automne) à partir d’un mois et d’un jour.

**Code utilisé**


```
def saison(m, j):

    if (m == 12 and j >= 22) or m < 3 or (m == 3 and j < 20):
        return "hiver"
    elif (m == 3 and j >= 20) or 4 <= m <= 5 or (m == 6 and j < 21):
        return "printemps"
    elif (m == 6 and j >= 21) or 7 <= m <= 8 or (m == 9 and j < 23):
        return "été"
    else:
        return "automne"

# Tests
tests = [
    (12, 31), (3, 19), (3, 20), (5, 31),
    (6, 20), (6, 21), (8, 10), (9, 22),
    (9, 23), (10, 30), (12, 21), (12, 22)
]

# Affichage des résultats
for m, j in tests:
    print(f"{m} {j} ==> est : {saison(m, j)}")
```

# **Q8 : Affichage d’un intervalle d’entiers**


**Description**

Affichage de tous les entiers dans un intervalle donné, en ordre croissant.

**Code utilisé**

```
def afficher_intervalle(debut, fin):
    if debut <= fin:
        return " ".join(str(i) for i in range(debut, fin + 1))
    return ""  # Chaîne vide si intervalle vide

# Tests
tests = [(-10, 5), (-2, 1), (-7, -10)]

# Affichage des résultats
for debut, fin in tests:
    print(f"{debut} {fin} ==> {afficher_intervalle(debut, fin)}")
```

# **Q9 : Génération de tables de multiplication**


**Description**

Génération et affichage des tables de multiplication de 1 à 10.

**Code utilisé**

```
def generer_tables():
    lignes = []
    for i in range(1, 11):
        ligne = [str(i * j) for j in range(1, 11)]
        lignes.append(" ".join(ligne))
    return "\n".join(lignes)

# Affichage
print(generer_tables())
```

# **Q10 : Affichage des diviseurs d’un nombre**

**Description**

Affichage des diviseurs d’un nombre entier dans l’ordre décroissant, suivi du nombre total de diviseurs.



**Code utilisé**



```
def afficher_diviseurs(n):
    diviseurs = [i for i in range(1, n + 1) if n % i == 0]
    diviseurs.reverse()  # ordre décroissant

    print(f"{n} ===> ", end="")
    for d in diviseurs:
        print(d, end=" ")
    print(f"===> {len(diviseurs)}")

# Tests
nombres = [24, 15, 123, 423]
for n in nombres:
    afficher_diviseurs(n)
```
