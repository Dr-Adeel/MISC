### Exercice 1
Ici, on calcule simplement l'aire et le périmètre d'un cercle avec un rayon donné.

```python
import math

r = 4.56


aire = math.pi * (r * r)
perimetre = 2 * math.pi * r


print("ona r = {} alors aire = {:.2f} et perimetre  = {:.2f}".format(r,aire,perimetre))
```

### Exercice 2
Dans cet exercice, on additionne des fractions et on simplifie le résultat en cherchant le diviseur commun (PGCD).

```python
import math
from fractions import Fraction 


n1, d1 = 217, 440
n2, d2 = 101, 256
n3, d3 = 86, 71


num_inter = (n1 * d2) + (n2 * d1)
den_inter = d1 * d2


num_final = (num_inter * d3) + (n3 * den_inter)
den_final = den_inter * d3


diviseur_commun = math.gcd(num_final, den_final)
num_reduit = num_final // diviseur_commun
den_reduit = den_final // diviseur_commun

print(f"Résultat calculé : {num_reduit}/{den_reduit}")

verif = Fraction(217, 440) + Fraction(101, 256) + Fraction(86, 71)
print("Verification {}".format(verif))
```

### Exercice 4
On détermine la catégorie sportive d'un enfant en fonction de son âge.

```python
age = int(input("entrez l'age de l'enfant : "))

if 0 <= age <= 5:
    print("OUPS pas suffisamment grand") 
elif 6 <= age <= 9:
    print("poussin ou pupille") 
elif 10 <= age <= 15:
    print("minime ou cadet") 
elif 16 <= age < 18:
    print("junior") 
elif age >= 18:
    print("senior") 
```

### Exercice 6
On vérifie si le magasin est ouvert ou fermé en fonction du jour et de l'heure. C'est de la logique booléenne.

```python
jour = int(input("Jour (0=Dim...6=Sam) : "))
heure = int(input("Heure (0-24) : "))

est_semaine = (jour >= 1 and jour <= 5)
est_samedi = (jour == 6)
matin = (heure >= 8 and heure < 13)
apres_midi = (heure >= 14 and heure < 17)

ouvert = False

if est_semaine:
    if matin or apres_midi:
        ouvert = True
elif est_samedi:
    if matin: 
        ouvert = True

if ouvert:
    print("Vrai")
else:
    print("Faux")
```

### Exercice 7
Ce programme trouve la saison correspondante à une date donnée (jour et mois).

```python
m = int(input("Mois (1-12) : "))
j = int(input("Jour (1-31) : "))

saison = ""

if (m == 12 and j >= 22) or m == 1 or m == 2 or (m == 3 and j < 20):
    saison = "hiver"
elif (m == 3 and j >= 20) or m == 4 or m == 5 or (m == 6 and j < 21):
    saison = "printemps" 
elif (m == 6 and j >= 21) or m == 7 or m == 8 or (m == 9 and j < 23):
    saison = "ete" 
else:
    saison = "automne" 

print(saison)
```

### Exercice 10
On cherche tous les diviseurs d'un nombre et on les compte.

```python
nombre = int(input("sisair votre nombre postive : "))
liste = []
compteur = 0 
for i in range(nombre , 0 , -1):
    if nombre % i == 0:
        liste.append(i)
        compteur = compteur + 1

print("le nombre {} ==> {} ==> {}".format(nombre , liste , compteur))
```
